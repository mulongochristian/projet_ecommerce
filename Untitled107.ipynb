```python
# ==========================================================
# PARSING MULTI-FORMATS DES TICKETS (corrigé)
# ==========================================================

import re
import pandas as pd

with open(TXT_FILE, "r", encoding="utf-8", errors="ignore") as f:
    raw_text = f.read()

# 1. Découpage sur le séparateur réel entre tickets
blocks = raw_text.split("=" * 40)
blocks = [b.strip() for b in blocks if b.strip()]
print(f"Nombre de blocs détectés : {len(blocks)}")

# 2. Patterns pour chaque format d'en-tête observé
PATTERNS = [
    # Format 1 : From: CUST_x <email> ... Subject: Ticket TKT_x - Order ORD_x
    re.compile(
        r"From:\s*(?P<cust>CUST_\d+)\s*<(?P<email>[^>]+)>.*?"
        r"Date:\s*(?P<date>.+?)\n.*?"
        r"Subject:.*?(?P<ticket>TKT_\d+).*?(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 2 : Ticket ID: TKT_x | Customer: CUST_x | Date: ... \n Order: ORD_x
    re.compile(
        r"Ticket ID:\s*(?P<ticket>TKT_\d+)\s*\|\s*Customer:\s*(?P<cust>CUST_\d+)\s*\|\s*Date:\s*(?P<date>.+?)\n"
        r".*?Order:\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 3 : [date] Ticket #TKT_x | Customer CUST_x | Order ORD_x
    re.compile(
        r"\[(?P<date>[^\]]+)\]\s*Ticket #(?P<ticket>TKT_\d+)\s*\|\s*Customer\s*(?P<cust>CUST_\d+)\s*\|\s*Order\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 4 : De: CUST_x ... Date du message: ... Référence: TKT_x / ORD_x
    re.compile(
        r"De:\s*(?P<cust>CUST_\d+).*?"
        r"Date du message:\s*(?P<date>.+?)\n.*?"
        r"Référence:\s*(?P<ticket>TKT_\d+)\s*/\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 5 : === TICKET TKT_x === \n Client: CUST_x \n Commande: ORD_x
    re.compile(
        r"===\s*TICKET\s*(?P<ticket>TKT_\d+)\s*===.*?"
        r"Client:\s*(?P<cust>CUST_\d+).*?"
        r"Commande:\s*(?P<order>ORD_\d+)",
        re.S
    ),
]

def parse_block(block):
    for pat in PATTERNS:
        m = pat.search(block)
        if m:
            gd = m.groupdict()
            # le corps = tout ce qui suit la fin du header matché
            body = block[m.end():].strip()
            # nettoyage du séparateur "---" éventuel en tête de corps
            body = re.sub(r"^-{2,}\s*", "", body).strip()
            return {
                "ticket_id": gd.get("ticket"),
                "customer_id": gd.get("cust"),
                "order_id": gd.get("order"),
                "date_raw": gd.get("date"),
                "email": gd.get("email"),
                "message": body,
            }
    # aucun pattern reconnu -> on garde le bloc brut pour investigation
    return {
        "ticket_id": None, "customer_id": None, "order_id": None,
        "date_raw": None, "email": None, "message": block
    }

parsed = [parse_block(b) for b in blocks]
df_tickets = pd.DataFrame(parsed)

print(f"Tickets parsés : {len(df_tickets)}")
print(f"Tickets non reconnus (ticket_id NaN) : {df_tickets['ticket_id'].isnull().sum()}")
display(df_tickets.head(10))
```

```python
%%writefile /content/projet_ecommerce/src/cleaning/clean_dates.py
# ==========================================================
# src/cleaning/clean_dates.py
# Parseur de dates universel — gère les formats mixtes
# ISO, US, FR, texte anglais, avec heure optionnelle
# ==========================================================

import re
import pandas as pd
from datetime import datetime

DATE_FORMATS = [
    "%Y-%m-%d %H:%M:%S",
    "%Y-%m-%d",
    "%Y/%m/%d",
    "%d-%m-%Y %H:%M",
    "%d-%m-%Y",
    "%d/%m/%Y %H:%M",
    "%m/%d/%Y",
    "%d/%m/%Y",
    "%B %d, %Y",
    "%d %B %Y",
    "%B %d %Y",
]


def _try_parse(date_str, fmt):
    try:
        return datetime.strptime(date_str.strip(), fmt)
    except (ValueError, AttributeError):
        return None


def parse_date_smart(date_str):
    """
    Retourne un dict:
      - parsed_date: datetime ou None
      - confidence: 'high' | 'ambiguous' | 'failed'
      - original: la valeur brute
    """
    if pd.isna(date_str) or not str(date_str).strip():
        return {"parsed_date": None, "confidence": "failed", "original": date_str}

    s = str(date_str).strip()

    unambiguous_formats = [
        "%Y-%m-%d %H:%M:%S", "%Y-%m-%d", "%Y/%m/%d",
        "%B %d, %Y", "%d %B %Y", "%B %d %Y",
    ]
    for fmt in unambiguous_formats:
        dt = _try_parse(s, fmt)
        if dt:
            return {"parsed_date": dt, "confidence": "high", "original": s}

    m = re.match(r"^(\d{1,2})[/-](\d{1,2})[/-](\d{4})(?:\s+(\d{1,2}):(\d{2}))?$", s)
    if m:
        a, b, year, hh, mm = m.groups()
        a, b, year = int(a), int(b), int(year)
        hh = int(hh) if hh else 0
        mm = int(mm) if mm else 0

        a_valid_month = 1 <= a <= 12
        b_valid_month = 1 <= b <= 12

        try:
            if a > 12 and b_valid_month:
                dt = datetime(year, b, a, hh, mm)
                return {"parsed_date": dt, "confidence": "high", "original": s}
            elif b > 12 and a_valid_month:
                dt = datetime(year, a, b, hh, mm)
                return {"parsed_date": dt, "confidence": "high", "original": s}
            elif a <= 12 and b <= 12:
                dt = datetime(year, b, a, hh, mm)
                return {"parsed_date": dt, "confidence": "ambiguous", "original": s}
        except ValueError:
            pass

    return {"parsed_date": None, "confidence": "failed", "original": s}


def clean_date_column(df, column, new_column=None):
    """
    Applique parse_date_smart sur une colonne complète et ajoute
    les colonnes <new_column>, <new_column>_confidence.
    """
    new_column = new_column or f"{column}_clean"
    results = df[column].apply(parse_date_smart)

    df[new_column] = results.apply(lambda r: r["parsed_date"])
    df[f"{new_column}_confidence"] = results.apply(lambda r: r["confidence"])

    report = df[f"{new_column}_confidence"].value_counts().to_dict()
    total = len(df)
    print(f"[clean_date_column] '{column}' -> '{new_column}'")
    print(f"  high      : {report.get('high', 0):>7} ({report.get('high', 0)/total*100:.2f}%)")
    print(f"  ambiguous : {report.get('ambiguous', 0):>7} ({report.get('ambiguous', 0)/total*100:.2f}%)")
    print(f"  failed    : {report.get('failed', 0):>7} ({report.get('failed', 0)/total*100:.2f}%)")

    return df, report
```

```python
%%writefile /content/projet_ecommerce/src/cleaning/clean_sales.py
# ==========================================================
# src/cleaning/clean_sales.py
# Nettoyage complet de sales_history :
# country, quantity, unit_price, doublons
# ==========================================================

import re
import pandas as pd
import numpy as np

# Toutes les variantes observées pour la France -> code ISO FR
COUNTRY_MAPPING = {
    "FRA": "FR", "FRANCE": "FR", "France": "FR", "france": "FR",
    "33": "FR", "F": "FR", "🇫🇷": "FR", "fr": "FR", "FR": "FR",
    "IT": "IT", "ES": "ES", "DE": "DE", "US": "US", "UK": "UK",
}

# Bornes de plausibilité pour quantity (basées sur la distribution observée : 1-9)
QUANTITY_MIN_VALID = 1
QUANTITY_MAX_VALID = 20  # au-delà, considéré aberrant (ex: 50000, 99999)


def normalize_country(df, column="country", log=None):
    """Normalise toutes les variantes de pays vers un code ISO unique."""
    before_unique = df[column].nunique(dropna=True)
    df[column] = df[column].map(COUNTRY_MAPPING).fillna(df[column])
    after_unique = df[column].nunique(dropna=True)

    msg = f"[normalize_country] '{column}': {before_unique} valeurs uniques -> {after_unique}"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def _quantity_to_numeric(val):
    """Convertit 'deux' -> NaN (traité comme aberrant, cf. décision projet),
    et force tout le reste en numérique."""
    if pd.isna(val):
        return np.nan
    try:
        return float(val)
    except (ValueError, TypeError):
        return np.nan  # ex: "deux" -> NaN, sera imputé comme les autres aberrants


def clean_quantity(df, column="quantity", log=None):
    """
    Valeurs aberrantes (négatives, 0, hors plage, non numériques comme 'deux')
    -> NaN -> imputation par la médiane des valeurs valides.
    """
    df[column] = df[column].apply(_quantity_to_numeric)

    valid_mask = (df[column] >= QUANTITY_MIN_VALID) & (df[column] <= QUANTITY_MAX_VALID)
    n_aberrant = (~valid_mask & df[column].notna()).sum()
    n_missing_native = df[column].isna().sum()

    median_valid = df.loc[valid_mask, column].median()

    df.loc[~valid_mask, column] = np.nan
    df[column] = df[column].fillna(median_valid)
    df[column] = df[column].astype(int)

    msg = (f"[clean_quantity] {n_aberrant} valeurs aberrantes + "
           f"{n_missing_native} manquantes natives -> imputées par médiane={median_valid}")
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_unit_price(df, column="unit_price", log=None):
    """Conversion en float, les non-convertibles deviennent NaN (à imputer par médiane)."""
    n_before_na = df[column].isna().sum()
    df[column] = pd.to_numeric(df[column], errors="coerce")
    n_after_na = df[column].isna().sum()
    n_new_na = n_after_na - n_before_na

    median_price = df[column].median()
    df[column] = df[column].fillna(median_price)

    msg = (f"[clean_unit_price] {n_new_na} valeurs non convertibles détectées, "
           f"total {n_after_na} imputées par médiane={median_price:.2f}")
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def remove_duplicates(df, log=None):
    """Supprime les doublons exacts (toutes colonnes identiques)."""
    n_before = len(df)
    df = df.drop_duplicates().reset_index(drop=True)
    n_removed = n_before - len(df)

    msg = f"[remove_duplicates] {n_removed} doublons exacts supprimés ({n_before} -> {len(df)})"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_sales_pipeline(df):
    """
    Pipeline complet de nettoyage sales_history.
    Retourne (df_nettoyé, log_transformations).
    Note : ne touche pas à transaction_date_clean, déjà traité par clean_dates.py.
    """
    log = []
    print("="*70)
    print("PIPELINE DE NETTOYAGE — SALES_HISTORY")
    print("="*70)

    df = remove_duplicates(df, log=log)
    df = normalize_country(df, log=log)
    df = clean_quantity(df, log=log)
    df = clean_unit_price(df, log=log)

    print("="*70)
    print(f"Terminé. Shape finale : {df.shape}")
    return df, log
```

```python
%%writefile /content/projet_ecommerce/src/cleaning/clean_reviews.py
# ==========================================================
# src/cleaning/clean_reviews.py
# Nettoyage complet de customer_reviews :
# rating, review_text (HTML), metadata (flatten), is_verified_purchase
# ==========================================================

import re
import html
import pandas as pd
import numpy as np

RATING_TEXT_MAPPING = {
    "nul": 1,
    "moyen": 3,
    "pas mal": 4,
    "excellent": 5,
    "une étoile": 1,
    "deux étoiles": 2,
    "trois étoiles": 3,
    "quatre étoiles": 4,
    "cinq étoiles": 5,
}

STAR_CHARS = {"⭐", "★"}


def _normalize_rating_value(val):
    if pd.isna(val):
        return np.nan

    s = str(val).strip()

    if re.fullmatch(r"\d+", s):
        n = int(s)
        return n if 1 <= n <= 5 else np.nan

    m = re.fullmatch(r"(\d+)\s*/\s*(\d+)", s)
    if m:
        num, den = int(m.group(1)), int(m.group(2))
        if den > 0:
            scaled = round(num / den * 5)
            return min(max(scaled, 1), 5)

    if s and all(c in STAR_CHARS for c in s):
        return min(max(len(s), 1), 5)

    s_lower = s.lower()
    if s_lower in RATING_TEXT_MAPPING:
        return RATING_TEXT_MAPPING[s_lower]

    return np.nan


def clean_rating(df, column="rating", log=None):
    df["rating_original"] = df[column]
    df[column] = df[column].apply(_normalize_rating_value)

    n_unmapped = df[column].isna().sum()
    median_rating = df[column].median()
    df[column] = df[column].fillna(median_rating).astype(int)

    msg = f"[clean_rating] {n_unmapped} valeurs non reconnues -> imputées par médiane={median_rating}"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_review_text(df, column="review_text", log=None):
    n_with_html = df[column].astype(str).str.contains(r"<[^>]+>", regex=True, na=False).sum()

    def strip_html(text):
        if pd.isna(text):
            return text
        text = re.sub(r"<[^>]+>", " ", text)
        text = html.unescape(text)
        text = re.sub(r"\s+", " ", text).strip()
        return text

    df[column] = df[column].apply(strip_html)

    msg = f"[clean_review_text] {n_with_html} textes contenant du HTML nettoyés"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def flatten_metadata(df, column="metadata", log=None):
    def safe_get(d, *keys):
        cur = d
        for k in keys:
            if not isinstance(cur, dict):
                return np.nan
            cur = cur.get(k)
        return cur if cur is not None else np.nan

    n_missing = df[column].isna().sum()

    df["metadata_country"] = df[column].apply(lambda d: safe_get(d, "user", "location", "country", "code"))
    df["metadata_city"] = df[column].apply(lambda d: safe_get(d, "user", "location", "city"))
    df["metadata_account_age_days"] = df[column].apply(lambda d: safe_get(d, "user", "account_age_days"))
    df["metadata_total_reviews"] = df[column].apply(lambda d: safe_get(d, "user", "total_reviews"))
    df["metadata_helpful_votes"] = df[column].apply(lambda d: safe_get(d, "helpful_votes"))
    df["metadata_platform"] = df[column].apply(lambda d: safe_get(d, "platform"))

    df = df.drop(columns=[column])

    msg = f"[flatten_metadata] 6 colonnes extraites, {n_missing} metadata manquantes d'origine (colonnes filles restent NaN)"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_is_verified_purchase(df, column="is_verified_purchase", log=None):
    n_missing = df[column].isna().sum()
    mode_value = df[column].mode(dropna=True)
    mode_value = mode_value.iloc[0] if not mode_value.empty else False

    df[column] = df[column].fillna(mode_value)

    msg = f"[clean_is_verified_purchase] {n_missing} valeurs manquantes -> imputées par mode={mode_value}"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_reviews_pipeline(df):
    log = []
    print("="*70)
    print("PIPELINE DE NETTOYAGE — CUSTOMER_REVIEWS")
    print("="*70)

    df = clean_rating(df, log=log)
    df = clean_review_text(df, log=log)
    df = flatten_metadata(df, log=log)
    df = clean_is_verified_purchase(df, log=log)

    print("="*70)
    print(f"Terminé. Shape finale : {df.shape}")
    return df, log
```

```python
%%writefile /content/projet_ecommerce/src/cleaning/clean_tickets.py
# ==========================================================
# src/cleaning/clean_tickets.py
# Nettoyage de support_tickets :
# normalisation espaces, suppression signatures, gestion NaN
# NOTE: la casse (MAJUSCULES) est préservée volontairement,
# c'est un signal potentiel d'urgence pour la Phase 6 (ML)
# ==========================================================

import re
import pandas as pd
import numpy as np

# Motifs de signature à retirer (fin de message)
SIGNATURE_PATTERNS = [
    r"--\s*Cordialement,?\s*\w*",
    r"--\s*Best regards,?\s*\w*",
    r"Merci\.?\s*Thank you very much!?\s*/?\s*Merci beaucoup!?",
]


def clean_message_text(df, column="message", log=None):
    """
    Nettoie le texte libre des tickets :
    - normalise les espaces multiples (mais garde la casse)
    - retire les signatures standardisées
    - retire les caractères de contrôle résiduels
    """
    n_before_empty = df[column].isna().sum()

    def clean_one(text):
        if pd.isna(text):
            return text
        # Retrait des signatures connues
        for pattern in SIGNATURE_PATTERNS:
            text = re.sub(pattern, "", text, flags=re.IGNORECASE)
        # Normalisation des espaces multiples / tabulations / retours ligne excessifs
        text = re.sub(r"[ \t]+", " ", text)
        text = re.sub(r"\n{2,}", "\n", text)
        text = text.strip()
        return text

    df[column] = df[column].apply(clean_one)

    n_empty_after = (df[column].astype(str).str.strip() == "").sum()

    msg = f"[clean_message_text] {n_before_empty} messages NaN natifs, {n_empty_after} vides après nettoyage"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def normalize_ids_case(df, log=None):
    """
    Force ticket_id, customer_id, order_id en majuscules
    (cohérence pour les futures jointures Phase 3).
    """
    for col in ["ticket_id", "customer_id", "order_id"]:
        if col in df.columns:
            n_lower = df[col].astype(str).str.islower().sum()
            df[col] = df[col].astype(str).str.upper()

    msg = "[normalize_ids_case] ticket_id/customer_id/order_id forcés en majuscules"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def flag_missing_email(df, column="email", log=None):
    """
    email absent dans 4 formats sur 5 (comportement attendu, pas une anomalie).
    On ajoute juste un flag explicite plutôt que d'imputer une fausse valeur.
    """
    n_missing = df[column].isna().sum()
    df["has_email"] = df[column].notna()

    msg = f"[flag_missing_email] {n_missing} emails absents ({n_missing/len(df)*100:.1f}%) -- attendu, flag 'has_email' créé"
    print(msg)
    if log is not None:
        log.append(msg)
    return df


def clean_tickets_pipeline(df):
    """
    Pipeline complet de nettoyage support_tickets.
    Note : ne touche pas à date_clean, déjà traité par clean_dates.py.
    """
    log = []
    print("="*70)
    print("PIPELINE DE NETTOYAGE — SUPPORT_TICKETS")
    print("="*70)

    df = normalize_ids_case(df, log=log)
    df = clean_message_text(df, log=log)
    df = flag_missing_email(df, log=log)

    print("="*70)
    print(f"Terminé. Shape finale : {df.shape}")
    return df, log
```

```python
# ==========================================================
# RE-PARSING COMPLET DES TICKETS (corrigé : mojibake + format 5)
# ==========================================================

import re
import ftfy
import pandas as pd

with open(TXT_FILE, "r", encoding="utf-8", errors="ignore") as f:
    raw_text = f.read()

# Correction du mojibake AVANT tout découpage (ftfy gère le texte mixte
# emoji + accents corrompus sans casser les emojis)
raw_text = ftfy.fix_text(raw_text)

blocks = raw_text.split("=" * 40)
blocks = [b.strip() for b in blocks if b.strip()]
print(f"Nombre de blocs détectés : {len(blocks)}")

PATTERNS = [
    # Format 1
    re.compile(
        r"From:\s*(?P<cust>CUST_\d+)\s*<(?P<email>[^>]+)>.*?"
        r"Date:\s*(?P<date>.+?)\n.*?"
        r"Subject:.*?(?P<ticket>TKT_\d+).*?(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 2
    re.compile(
        r"Ticket ID:\s*(?P<ticket>TKT_\d+)\s*\|\s*Customer:\s*(?P<cust>CUST_\d+)\s*\|\s*Date:\s*(?P<date>.+?)\n"
        r".*?Order:\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 3
    re.compile(
        r"\[(?P<date>[^\]]+)\]\s*Ticket #(?P<ticket>TKT_\d+)\s*\|\s*Customer\s*(?P<cust>CUST_\d+)\s*\|\s*Order\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 4
    re.compile(
        r"De:\s*(?P<cust>CUST_\d+).*?"
        r"Date du message:\s*(?P<date>.+?)\n.*?"
        r"Référence:\s*(?P<ticket>TKT_\d+)\s*/\s*(?P<order>ORD_\d+)",
        re.S
    ),
    # Format 5 CORRIGÉ : consomme désormais le "Date:" optionnel ET le "===" final,
    # donc plus de fuite dans le corps du message
    re.compile(
        r"===\s*TICKET\s*(?P<ticket>TKT_\d+)\s*===\s*\n"
        r"Client:\s*(?P<cust>CUST_\d+)\s*\n"
        r"Commande:\s*(?P<order>ORD_\d+)\s*\n"
        r"(?:Date:\s*(?P<date>.+?)\s*\n)?"
        r"==="
    ),
]

def parse_block(block):
    for pat in PATTERNS:
        m = pat.search(block)
        if m:
            gd = m.groupdict()
            body = block[m.end():].strip()
            body = re.sub(r"^-{2,}\s*", "", body).strip()
            return {
                "ticket_id": gd.get("ticket"),
                "customer_id": gd.get("cust"),
                "order_id": gd.get("order"),
                "date_raw": gd.get("date"),
                "email": gd.get("email"),
                "message": body,
            }
    return {
        "ticket_id": None, "customer_id": None, "order_id": None,
        "date_raw": None, "email": None, "message": block
    }

parsed = [parse_block(b) for b in blocks]
df_tickets = pd.DataFrame(parsed)

print(f"Tickets parsés : {len(df_tickets)}")
print(f"Tickets non reconnus (ticket_id NaN) : {df_tickets['ticket_id'].isnull().sum()}")
display(df_tickets.head(10))
```

```python
%%writefile /content/projet_ecommerce/src/transform/build_dimensions.py
# ==========================================================
# src/transform/build_dimensions.py
# Construction des dimensions du schéma en étoile
# ==========================================================

import pandas as pd


def build_dim_customers(df_sales, df_reviews, df_tickets):
    """
    Dimension client : union des customer_id vus dans les 3 sources.
    Le pays est pris depuis sales (source la plus fiable pour ce champ).
    """
    cust_sales = df_sales[["customer_id", "country"]].drop_duplicates("customer_id")
    all_ids = pd.concat([
        df_sales["customer_id"],
        df_reviews["user_id"].rename("customer_id"),
        df_tickets["customer_id"],
    ]).dropna().unique()

    dim_customers = pd.DataFrame({"customer_id": all_ids})
    dim_customers = dim_customers.merge(cust_sales, on="customer_id", how="left")
    dim_customers["country"] = dim_customers["country"].fillna("UNKNOWN")

    print(f"[dim_customers] {len(dim_customers)} clients uniques (sales+reviews+tickets)")
    return dim_customers


def build_dim_products(df_sales, df_reviews):
    """Dimension produit : union des product_id vus dans sales et reviews."""
    all_products = pd.concat([
        df_sales["product_id"],
        df_reviews["product_id"],
    ]).dropna().unique()

    dim_products = pd.DataFrame({"product_id": sorted(all_products)})

    print(f"[dim_products] {len(dim_products)} produits uniques (sales+reviews)")
    return dim_products


def build_dim_time(df_sales, df_reviews, df_tickets):
    """
    Dimension temps : plage complète couvrant les 3 sources,
    granularité journalière, avec attributs calendaires utiles.
    """
    all_dates = pd.concat([
        df_sales["transaction_date_clean"],
        df_reviews["review_date_clean"],
        df_tickets["date_clean"],
    ]).dropna()

    date_min, date_max = all_dates.min().normalize(), all_dates.max().normalize()
    dim_time = pd.DataFrame({"date": pd.date_range(date_min, date_max, freq="D")})
    dim_time["year"] = dim_time["date"].dt.year
    dim_time["month"] = dim_time["date"].dt.month
    dim_time["day"] = dim_time["date"].dt.day
    dim_time["day_of_week"] = dim_time["date"].dt.dayofweek
    dim_time["is_weekend"] = dim_time["day_of_week"].isin([5, 6])

    print(f"[dim_time] {len(dim_time)} jours ({date_min.date()} -> {date_max.date()})")
    return dim_time
```

```python
%%writefile /content/projet_ecommerce/src/transform/build_facts.py
# ==========================================================
# src/transform/build_facts.py
# Construction des tables de faits du schéma en étoile
# fact_sales, fact_reviews, fact_tickets
#
# HYPOTHÈSE MÉTHODOLOGIQUE DOCUMENTÉE :
# Le order_id cité dans les tickets ne correspond à aucun
# transaction_id réel de sales (0% de correspondance directe).
# Stratégie retenue : jointure tickets->sales via customer_id,
# en associant à chaque ticket la transaction du même client
# la plus proche dans le temps (avant la date du ticket).
# Les tickets sans aucune transaction du client correspondant
# sont conservés dans fact_tickets mais exclus des features
# produit (ratio tickets/ventes).
# ==========================================================

import pandas as pd
import numpy as np


def build_fact_sales(df_sales):
    """Table de faits ventes : une ligne par transaction."""
    fact_sales = df_sales[[
        "transaction_id", "customer_id", "product_id",
        "quantity", "unit_price", "country", "payment_method",
        "transaction_date_clean"
    ]].copy()
    fact_sales["revenue"] = fact_sales["quantity"] * fact_sales["unit_price"]

    print(f"[fact_sales] {len(fact_sales)} lignes, CA total = {fact_sales['revenue'].sum():,.2f}")
    return fact_sales


def build_fact_reviews(df_reviews):
    """Table de faits avis : une ligne par avis."""
    fact_reviews = df_reviews[[
        "review_id", "user_id", "product_id", "rating",
        "review_text", "is_verified_purchase", "review_date_clean",
        "metadata_account_age_days", "metadata_helpful_votes"
    ]].copy()
    fact_reviews = fact_reviews.rename(columns={"user_id": "customer_id"})

    print(f"[fact_reviews] {len(fact_reviews)} lignes, note moyenne = {fact_reviews['rating'].mean():.2f}")
    return fact_reviews


def link_ticket_to_nearest_sale(df_tickets, df_sales):
    """
    Associe chaque ticket à la transaction du même customer_id
    la plus proche temporellement AVANT la date du ticket.
    Utilise merge_asof (jointure temporelle native pandas, efficace
    même sur de gros volumes).
    """
    tickets_sorted = df_tickets.dropna(subset=["date_clean"]).sort_values("date_clean").copy()
    sales_sorted = df_sales.dropna(subset=["transaction_date_clean"]).sort_values("transaction_date_clean").copy()

    linked_parts = []
    n_customers_matched = 0

    # merge_asof exige un tri global ET par groupe -> on boucle par customer
    # pour garantir l'exactitude (le volume par client reste faible, <50 lignes)
    for cust_id, ticket_group in tickets_sorted.groupby("customer_id"):
        cust_sales = sales_sorted[sales_sorted["customer_id"] == cust_id]
        if cust_sales.empty:
            linked = ticket_group.copy()
            linked["matched_product_id"] = np.nan
            linked["matched_transaction_id"] = np.nan
            linked["days_since_purchase"] = np.nan
        else:
            n_customers_matched += 1
            linked = pd.merge_asof(
                ticket_group.sort_values("date_clean"),
                cust_sales[["transaction_id", "product_id", "transaction_date_clean"]].sort_values("transaction_date_clean"),
                left_on="date_clean",
                right_on="transaction_date_clean",
                direction="backward",
            )
            linked["days_since_purchase"] = (
                linked["date_clean"] - linked["transaction_date_clean"]
            ).dt.days
            linked = linked.rename(columns={
                "product_id": "matched_product_id",
                "transaction_id": "matched_transaction_id",
            })
        linked_parts.append(linked)

    fact_tickets = pd.concat(linked_parts, ignore_index=True)
    n_unmatched = fact_tickets["matched_product_id"].isna().sum()

    print(f"[link_ticket_to_nearest_sale] {n_customers_matched} clients avec au moins 1 vente correspondante")
    print(f"[link_ticket_to_nearest_sale] {n_unmatched} tickets sans transaction client correspondante "
          f"({n_unmatched/len(fact_tickets)*100:.2f}%) -- exclus des features produit")
    return fact_tickets


def build_fact_tickets(df_tickets, df_sales):
    """Table de faits tickets, enrichie du produit/transaction associés (approximation)."""
    linked = link_ticket_to_nearest_sale(df_tickets, df_sales)

    fact_tickets = linked[[
        "ticket_id", "customer_id", "order_id", "matched_transaction_id",
        "matched_product_id", "message", "date_clean",
        "has_email", "days_since_purchase"
    ]].copy()

    print(f"[fact_tickets] {len(fact_tickets)} lignes")
    return fact_tickets
```

# Étape suivante du projet : Features dérivées (toujours Phase 3)
Le cahier des charges demande explicitement 3 features :

Score de sentiment par produit


Ratio tickets/ventes par produit



Délai moyen entre achat et ticket support

```python
%%writefile /content/projet_ecommerce/src/transform/build_features.py
# ==========================================================
# src/transform/build_features.py
# Features dérivées par produit (Phase 3, cahier des charges) :
# 1. Score de sentiment (proxy rating normalisé -1/+1)
# 2. Ratio tickets/ventes
# 3. Délai moyen achat -> ticket
# ==========================================================

import pandas as pd
import numpy as np


def compute_sentiment_score(fact_reviews):
    """
    Score de sentiment par produit = moyenne des rating, normalisée sur [-1, +1].
    Formule: (rating_moyen - 3) / 2  (3 = neutre sur l'échelle 1-5)
    """
    agg = fact_reviews.groupby("product_id").agg(
        avg_rating=("rating", "mean"),
        n_reviews=("rating", "count"),
    ).reset_index()

    agg["sentiment_score"] = (agg["avg_rating"] - 3) / 2

    print(f"[compute_sentiment_score] {len(agg)} produits scorés, "
          f"sentiment moyen global = {agg['sentiment_score'].mean():.3f}")
    return agg[["product_id", "avg_rating", "n_reviews", "sentiment_score"]]


def compute_ticket_ratio(fact_sales, fact_tickets):
    """
    Ratio tickets/ventes par produit = nb tickets liés / nb ventes du produit.
    Les tickets orphelins (matched_product_id NaN) sont exclus, comme décidé.
    """
    sales_per_product = fact_sales.groupby("product_id").size().rename("n_sales")

    valid_tickets = fact_tickets.dropna(subset=["matched_product_id"])
    tickets_per_product = valid_tickets.groupby("matched_product_id").size().rename("n_tickets")
    tickets_per_product.index.name = "product_id"

    ratio_df = pd.concat([sales_per_product, tickets_per_product], axis=1).fillna(0)
    ratio_df["ticket_ratio"] = ratio_df["n_tickets"] / ratio_df["n_sales"].replace(0, np.nan)
    ratio_df["ticket_ratio"] = ratio_df["ticket_ratio"].fillna(0)
    ratio_df = ratio_df.reset_index()

    n_excluded = fact_tickets["matched_product_id"].isna().sum()
    print(f"[compute_ticket_ratio] {len(ratio_df)} produits, "
          f"{n_excluded} tickets orphelins exclus du calcul")
    return ratio_df[["product_id", "n_sales", "n_tickets", "ticket_ratio"]]


def compute_avg_delay(fact_tickets):
    """
    Délai moyen entre achat et ticket, par produit (matched_product_id).
    Exclut les tickets orphelins et les délais négatifs (incohérence temporelle).
    """
    valid = fact_tickets.dropna(subset=["matched_product_id", "days_since_purchase"])
    n_negative = (valid["days_since_purchase"] < 0).sum()
    valid = valid[valid["days_since_purchase"] >= 0]

    agg = valid.groupby("matched_product_id")["days_since_purchase"].mean().reset_index()
    agg = agg.rename(columns={"matched_product_id": "product_id", "days_since_purchase": "avg_delay_days"})

    print(f"[compute_avg_delay] {len(agg)} produits, {n_negative} délais négatifs exclus, "
          f"délai moyen global = {valid['days_since_purchase'].mean():.1f} jours")
    return agg


def build_product_features(dim_products, fact_sales, fact_reviews, fact_tickets):
    """Assemble les 3 features dérivées en une seule table produit enrichie."""
    sentiment = compute_sentiment_score(fact_reviews)
    ratio = compute_ticket_ratio(fact_sales, fact_tickets)
    delay = compute_avg_delay(fact_tickets)

    product_features = dim_products.merge(sentiment, on="product_id", how="left")
    product_features = product_features.merge(ratio, on="product_id", how="left")
    product_features = product_features.merge(delay, on="product_id", how="left")

    # Produits sans avis/ventes/tickets -> valeurs par défaut explicites (pas de biais silencieux)
    product_features["n_reviews"] = product_features["n_reviews"].fillna(0)
    product_features["sentiment_score"] = product_features["sentiment_score"].fillna(0)
    product_features["n_sales"] = product_features["n_sales"].fillna(0)
    product_features["n_tickets"] = product_features["n_tickets"].fillna(0)
    product_features["ticket_ratio"] = product_features["ticket_ratio"].fillna(0)
    # avg_delay_days reste NaN si aucun ticket -> pas de valeur par défaut arbitraire (pas de "0 jour" trompeur)

    print(f"[build_product_features] Table finale : {product_features.shape}")
    return product_features
```

# Dernière étape de la Phase 3 : Stockage Parquet Bronze → Silver → Gold

```python
%%writefile /content/projet_ecommerce/src/transform/save_parquet_layers.py
# ==========================================================
# src/transform/save_parquet_layers.py
# Architecture de stockage Bronze -> Silver -> Gold (Parquet)
# ==========================================================

import os
import pandas as pd


def save_bronze(df_sales_raw, df_reviews_raw, df_tickets_raw, base_path):
    """Couche Bronze : données brutes, aucune transformation."""
    os.makedirs(f"{base_path}/raw", exist_ok=True)

    # Convert mixed 'rating' column to string to avoid ArrowTypeError
    df_reviews_raw_copy = df_reviews_raw.copy()
    if 'rating' in df_reviews_raw_copy.columns:
        # More robust conversion to ensure all elements are pure strings
        df_reviews_raw_copy['rating'] = df_reviews_raw_copy['rating'].apply(lambda x: str(x) if pd.notna(x) else None)

    df_sales_raw.to_parquet(f"{base_path}/raw/sales_bronze.parquet", index=False)
    df_reviews_raw_copy.to_parquet(f"{base_path}/raw/reviews_bronze.parquet", index=False)
    df_tickets_raw.to_parquet(f"{base_path}/raw/tickets_bronze.parquet", index=False)

    print(f"[Bronze] sales={len(df_sales_raw)}, reviews={len(df_reviews_raw)}, tickets={len(df_tickets_raw)}")


def save_silver(df_sales_clean, df_reviews_clean, df_tickets_clean, base_path):
    """Couche Silver : données nettoyées (post src/cleaning/)."""
    os.makedirs(f"{base_path}/cleaned", exist_ok=True)

    # Robust conversion for 'rating_original' column in df_reviews_clean for Silver layer
    # This column can still contain mixed types (e.g., int, str) from the raw data
    # before cleaning, which PyArrow expects to be uniform for Parquet.
    df_reviews_clean_copy = df_reviews_clean.copy()
    if 'rating_original' in df_reviews_clean_copy.columns:
        # Create a new Series where every element is explicitly a string or None.
        # This bypasses any potential issues with in-place modifications or
        # complex interactions of apply/astype with existing object dtype.
        df_reviews_clean_copy['rating_original'] = pd.Series([
            str(x) if pd.notna(x) else None
            for x in df_reviews_clean_copy['rating_original']
        ], dtype=pd.StringDtype(), index=df_reviews_clean_copy.index)

    df_sales_clean.to_parquet(f"{base_path}/cleaned/sales_silver.parquet", index=False)
    df_reviews_clean_copy.to_parquet(f"{base_path}/cleaned/reviews_silver.parquet", index=False)
    df_tickets_clean.to_parquet(f"{base_path}/cleaned/tickets_silver.parquet", index=False)

    print(f"[Silver] sales={len(df_sales_clean)}, reviews={len(df_reviews_clean)}, tickets={len(df_tickets_clean)}")


def save_gold(dim_customers, dim_products, dim_time,
              fact_sales, fact_reviews, fact_tickets,
              product_features, base_path):
    """Couche Gold : dimensions, faits, et features -- prêt pour Phase 5/6."""
    os.makedirs(f"{base_path}/features", exist_ok=True)

    dim_customers.to_parquet(f"{base_path}/features/dim_customers.parquet", index=False)
    dim_products.to_parquet(f"{base_path}/features/dim_products.parquet", index=False)
    dim_time.to_parquet(f"{base_path}/features/dim_time.parquet", index=False)

    fact_sales.to_parquet(f"{base_path}/features/fact_sales.parquet", index=False)
    fact_reviews.to_parquet(f"{base_path}/features/fact_reviews.parquet", index=False)
    fact_tickets.to_parquet(f"{base_path}/features/fact_tickets.parquet", index=False)

    product_features.to_parquet(f"{base_path}/features/product_features_gold.parquet", index=False)

    print(f"[Gold] 3 dimensions, 3 faits, 1 table de features produit sauvegardés")
```

# Phase 4 — RGPD & Sécurité

```python
%%writefile /content/projet_ecommerce/src/rgpd/pseudonymization.py
# ==========================================================
# src/rgpd/pseudonymization.py
# 3 techniques de pseudonymisation (Phase 4A, cahier des charges) :
# 1. Chiffrement symétrique (Fernet) -- réversible, pour emails/IDs
# 2. Bruit différentiel (Laplace) -- pour métriques agrégées publiques
# 3. Masquage partiel -- pour affichage logs/rapports
# ==========================================================

import os
import numpy as np
import pandas as pd
from cryptography.fernet import Fernet

KEY_PATH = "/content/projet_ecommerce/.secrets/encryption.key"


# ----------------------------------------------------------
# 1. CHIFFREMENT SYMÉTRIQUE (réversible)
# ----------------------------------------------------------

def get_or_create_key(key_path=KEY_PATH):
    """
    Génère une clé de chiffrement une seule fois et la stocke localement.
    NOTE PRODUCTION: en environnement réel, cette clé irait dans un
    gestionnaire de secrets (AWS KMS, HashiCorp Vault), jamais sur disque
    en clair ni versionnée dans Git (.gitignore l'exclut ici).
    """
    os.makedirs(os.path.dirname(key_path), exist_ok=True)
    if os.path.exists(key_path):
        with open(key_path, "rb") as f:
            return f.read()
    key = Fernet.generate_key()
    with open(key_path, "wb") as f:
        f.write(key)
    print(f"[get_or_create_key] Nouvelle clé générée et stockée : {key_path}")
    return key


def encrypt_value(value, fernet):
    """Chiffre une valeur unique (str). Retourne None si NaN."""
    if pd.isna(value):
        return None
    return fernet.encrypt(str(value).encode()).decode()


def decrypt_value(token, fernet):
    """Déchiffre un token. Retourne None si NaN/invalide."""
    if pd.isna(token) or token is None:
        return None
    return fernet.decrypt(str(token).encode()).decode()


def encrypt_column(df, column, fernet, new_column=None):
    """Chiffre une colonne entière, conserve l'originale renommée '_original' à part (traçabilité dev only)."""
    new_column = new_column or f"{column}_encrypted"
    df[new_column] = df[column].apply(lambda v: encrypt_value(v, fernet))
    print(f"[encrypt_column] '{column}' -> '{new_column}' ({df[new_column].notna().sum()} valeurs chiffrées)")
    return df


# ----------------------------------------------------------
# 2. BRUIT DIFFÉRENTIEL (pour métriques agrégées)
# ----------------------------------------------------------

def add_laplace_noise(series, epsilon=1.0, sensitivity=1.0):
    """
    Ajoute un bruit de Laplace calibré (epsilon-differential privacy).
    epsilon plus petit = plus de bruit = plus de confidentialité, moins de précision.
    sensitivity = variation max qu'un individu peut causer sur la métrique.
    """
    scale = sensitivity / epsilon
    noise = np.random.laplace(loc=0, scale=scale, size=len(series))
    return series + noise


def apply_differential_privacy(df, columns, epsilon=1.0, log=None):
    """
    Applique le bruit différentiel sur des colonnes agrégées
    (ex: n_sales, n_reviews, n_tickets par produit) avant toute
    publication externe (dashboard Phase 5, rapport).
    Les comptages étant discrets, on arrondit et on empêche les négatifs.
    """
    df_noisy = df.copy()
    for col in columns:
        df_noisy[f"{col}_dp"] = add_laplace_noise(df[col], epsilon=epsilon, sensitivity=1.0)
        df_noisy[f"{col}_dp"] = df_noisy[f"{col}_dp"].clip(lower=0).round().astype(int)

    msg = f"[apply_differential_privacy] Bruit Laplace (epsilon={epsilon}) appliqué sur {columns}"
    print(msg)
    if log is not None:
        log.append(msg)
    return df_noisy


# ----------------------------------------------------------
# 3. MASQUAGE PARTIEL (affichage logs/rapports)
# ----------------------------------------------------------

def mask_email(email):
    """user1234@example.com -> u******4@example.com"""
    if pd.isna(email) or "@" not in str(email):
        return None
    local, domain = email.split("@", 1)
    if len(local) <= 2:
        masked_local = local[0] + "*" * (len(local) - 1)
    else:
        masked_local = local[0] + "*" * (len(local) - 2) + local[-1]
    return f"{masked_local}@{domain}"


def mask_id(identifier, keep_start=4, keep_end=2):
    """CUST_02735 -> CUST_****35"""
    if pd.isna(identifier):
        return None
    s = str(identifier)
    if len(s) <= keep_start + keep_end:
        return s[0] + "*" * (len(s) - 1)
    return s[:keep_start] + "*" * (len(s) - keep_start - keep_end) + s[-keep_end:]


def mask_column(df, column, mask_fn, new_column=None):
    new_column = new_column or f"{column}_masked"
    df[new_column] = df[column].apply(mask_fn)
    print(f"[mask_column] '{column}' -> '{new_column}' masqué pour {df[new_column].notna().sum()} valeurs")
    return df
```

# Étape suivante : Droit à l'oubli (Art. 17 RGPD) — Phase 4B

```python
%%writefile /content/projet_ecommerce/src/rgpd/right_to_erasure.py
# ==========================================================
# src/rgpd/right_to_erasure.py
# Droit à l'oubli (Art. 17 RGPD) — Phase 4B
# ==========================================================

import os
import json
import hashlib
import pandas as pd
from datetime import datetime, timezone

DATA_PATH = "/content/projet_ecommerce/data"
CERT_PATH = "/content/projet_ecommerce/rgpd/erasure_certificates"

# Mapping fichier -> colonnes contenant potentiellement l'identifiant client
LAYER_FILES = {
    "bronze": {
        "raw/sales_bronze.parquet": ["customer_id"],
        "raw/reviews_bronze.parquet": ["user_id"],
        # tickets_bronze est stocké en blocs texte bruts, traité à part (recherche texte)
    },
    "silver": {
        "cleaned/sales_silver.parquet": ["customer_id"],
        "cleaned/reviews_silver.parquet": ["user_id"],
        "cleaned/tickets_silver.parquet": ["customer_id"],
    },
    "gold": {
        "features/dim_customers.parquet": ["customer_id"],
        "features/fact_sales.parquet": ["customer_id"],
        "features/fact_reviews.parquet": ["customer_id"],
        "features/fact_tickets.parquet": ["customer_id"],
    },
}


def _certificate_hash(user_id, timestamp, details):
    """Empreinte d'intégrité du certificat (preuve que le contenu n'a pas été altéré après coup)."""
    payload = f"{user_id}|{timestamp}|{json.dumps(details, sort_keys=True)}"
    return hashlib.sha256(payload.encode()).hexdigest()


def _purge_from_file(filepath, columns, user_id):
    """Supprime toutes les lignes où l'une des colonnes == user_id. Retourne le nb de lignes supprimées."""
    if not os.path.exists(filepath):
        return 0, "file_not_found"

    df = pd.read_parquet(filepath)
    mask = pd.Series(False, index=df.index)
    for col in columns:
        if col in df.columns:
            mask |= (df[col] == user_id)

    n_removed = mask.sum()
    if n_removed > 0:
        df = df[~mask].reset_index(drop=True)
        df.to_parquet(filepath, index=False)

    return int(n_removed), "ok"


def _purge_tickets_bronze(user_id, base_path=DATA_PATH):
    """Cas spécifique : tickets_bronze stocke des blocs texte bruts, pas de colonne customer_id typée."""
    filepath = f"{base_path}/raw/tickets_bronze.parquet"
    if not os.path.exists(filepath):
        return 0, "file_not_found"

    df = pd.read_parquet(filepath)
    mask = df["raw_block"].astype(str).str.contains(user_id, regex=False)
    n_removed = mask.sum()
    if n_removed > 0:
        df = df[~mask].reset_index(drop=True)
        df.to_parquet(filepath, index=False)
    return int(n_removed), "ok"


def _purge_transformation_log(user_id, base_path=DATA_PATH):
    """Purge les logs (CSV/JSON) contenant l'identifiant en clair."""
    csv_path = f"{base_path}/cleaned/transformation_log.csv"
    json_path = f"{base_path}/cleaned/transformation_log.json"
    n_purged = 0

    if os.path.exists(csv_path):
        df_log = pd.read_csv(csv_path)
        mask = df_log["transformation"].astype(str).str.contains(user_id, regex=False)
        n_purged += mask.sum()
        if mask.any():
            df_log = df_log[~mask]
            df_log.to_csv(csv_path, index=False)

    if os.path.exists(json_path):
        with open(json_path, "r", encoding="utf-8") as f:
            entries = json.load(f)
        before = len(entries)
        entries = [e for e in entries if user_id not in json.dumps(e)]
        if len(entries) < before:
            with open(json_path, "w", encoding="utf-8") as f:
                json.dump(entries, f, ensure_ascii=False, indent=2)

    return int(n_purged)


def delete_user_data(user_id, base_path=DATA_PATH, cert_path=CERT_PATH):
    """
    Droit à l'oubli (Art. 17 RGPD) :
    - Supprime toutes les lignes liées à user_id dans Bronze, Silver, Gold
    - Purge les logs de transformation contenant l'identifiant
    - Génère un certificat de suppression horodaté et vérifiable
    """
    timestamp = datetime.now(timezone.utc).isoformat()
    details = {}

    for layer, files in LAYER_FILES.items():
        details[layer] = {}
        for rel_path, columns in files.items():
            full_path = f"{base_path}/{rel_path}"
            n_removed, status = _purge_from_file(full_path, columns, user_id)
            details[layer][rel_path] = {"rows_removed": n_removed, "status": status}

    n_tickets_bronze, status_tb = _purge_tickets_bronze(user_id, base_path)
    details["bronze"]["raw/tickets_bronze.parquet"] = {"rows_removed": n_tickets_bronze, "status": status_tb}

    n_logs_purged = _purge_transformation_log(user_id, base_path)
    details["logs"] = {"transformation_log": {"rows_removed": n_logs_purged, "status": "ok"}}

    total_removed = sum(
        f["rows_removed"] for layer in details.values() for f in layer.values()
    )

    certificate = {
        "user_id_hash": hashlib.sha256(user_id.encode()).hexdigest(),  # jamais l'ID en clair dans le certificat
        "timestamp_utc": timestamp,
        "total_rows_removed": int(total_removed),
        "details": details,
        "legal_basis": "Article 17 RGPD - Droit à l'effacement",
    }
    certificate["integrity_hash"] = _certificate_hash(user_id, timestamp, details)

    os.makedirs(cert_path, exist_ok=True)
    cert_filename = f"{cert_path}/erasure_{hashlib.sha256(user_id.encode()).hexdigest()[:12]}_{timestamp.replace(':', '-')}.json"
    with open(cert_filename, "w", encoding="utf-8") as f:
        json.dump(certificate, f, ensure_ascii=False, indent=2)

    print(f"[delete_user_data] user_id={user_id} : {total_removed} lignes supprimées au total")
    print(f"[delete_user_data] Certificat généré : {cert_filename}")

    return certificate
```

# Étape suivante : Droit à la portabilité (Art. 20 RGPD) — Phase 4C

```python
%%writefile /content/projet_ecommerce/src/rgpd/data_portability.py
# ==========================================================
# src/rgpd/data_portability.py
# Droit à la portabilité (Art. 20 RGPD) — Phase 4C
# ==========================================================

import os
import json
import pandas as pd
from datetime import datetime, timezone

DATA_PATH = "/content/projet_ecommerce/data"
EXPORT_PATH = "/content/projet_ecommerce/rgpd/data_exports"

# Finalité de traitement déclarée pour chaque source (cohérent avec le registre Art. 30)
PROCESSING_PURPOSES = {
    "sales": "Exécution du contrat de vente, facturation, statistiques commerciales",
    "reviews": "Amélioration produit, affichage public des avis, détection de fraude",
    "tickets": "Support client, résolution des réclamations, analyse de satisfaction",
}


def _extract_user_records(user_id, base_path=DATA_PATH):
    """Récupère toutes les lignes liées à user_id dans les couches Silver (données propres)."""
    records = {}

    sales_path = f"{base_path}/cleaned/sales_silver.parquet"
    if os.path.exists(sales_path):
        df = pd.read_parquet(sales_path)
        user_sales = df[df["customer_id"] == user_id]
        records["sales"] = user_sales.to_dict(orient="records")

    reviews_path = f"{base_path}/cleaned/reviews_silver.parquet"
    if os.path.exists(reviews_path):
        df = pd.read_parquet(reviews_path)
        user_reviews = df[df["user_id"] == user_id]
        records["reviews"] = user_reviews.to_dict(orient="records")

    tickets_path = f"{base_path}/cleaned/tickets_silver.parquet"
    if os.path.exists(tickets_path):
        df = pd.read_parquet(tickets_path)
        user_tickets = df[df["customer_id"] == user_id]
        records["tickets"] = user_tickets.to_dict(orient="records")

    return records


def _make_json_serializable(obj):
    """Convertit récursivement les types pandas/numpy non sérialisables (Timestamp, etc.)."""
    if isinstance(obj, dict):
        return {k: _make_json_serializable(v) for k, v in obj.items()}
    if isinstance(obj, list):
        return [_make_json_serializable(v) for v in obj]
    if isinstance(obj, (pd.Timestamp,)):
        return obj.isoformat()
    if pd.isna(obj) if not isinstance(obj, (list, dict)) else False:
        return None
    return obj


def export_user_data(user_id, format="json", base_path=DATA_PATH, export_path=EXPORT_PATH):
    """
    Droit à la portabilité (Art. 20 RGPD) :
    Extrait toutes les données d'un utilisateur dans un format structuré,
    avec métadonnées de collecte et finalité de traitement.
    """
    timestamp = datetime.now(timezone.utc).isoformat()
    records = _extract_user_records(user_id, base_path)

    n_total = sum(len(v) for v in records.values())

    export_package = {
        "export_metadata": {
            "user_id": user_id,
            "export_date_utc": timestamp,
            "legal_basis": "Article 20 RGPD - Droit à la portabilité des données",
            "format_version": "1.0",
            "total_records": n_total,
        },
        "data_sources": {
            source: {
                "purpose": PROCESSING_PURPOSES.get(source, "Non spécifié"),
                "record_count": len(data),
                "records": _make_json_serializable(data),
            }
            for source, data in records.items()
        },
    }

    os.makedirs(export_path, exist_ok=True)

    if format == "json":
        filename = f"{export_path}/export_{user_id}_{timestamp.replace(':', '-')}.json"
        with open(filename, "w", encoding="utf-8") as f:
            json.dump(export_package, f, ensure_ascii=False, indent=2)
    elif format == "csv":
        # Un CSV par source, dans un sous-dossier dédié à cet export
        subfolder = f"{export_path}/export_{user_id}_{timestamp.replace(':', '-')}"
        os.makedirs(subfolder, exist_ok=True)
        for source, data in records.items():
            if data:
                pd.DataFrame(data).to_csv(f"{subfolder}/{source}.csv", index=False)
        filename = subfolder
    else:
        raise ValueError(f"Format non supporté : {format} (attendu: 'json' ou 'csv')")

    print(f"[export_user_data] user_id={user_id} : {n_total} enregistrements exportés ({format})")
    print(f"[export_user_data] Fichier : {filename}")

    return export_package, filename
```

```python
%%writefile /content/projet_ecommerce/src/security/access_control.py
# ==========================================================
# src/security/access_control.py
# RBAC simulé (Role-Based Access Control) — Phase 4E
# ==========================================================

from datetime import datetime, timezone
from functools import wraps

# Définition des rôles et de leurs permissions sur les ressources
ROLE_PERMISSIONS = {
    "admin": {"sales", "reviews", "tickets", "pii", "delete", "export"},
    "data_scientist": {"sales", "reviews", "tickets"},  # pas d'accès direct aux PII brutes
    "support_agent": {"tickets", "pii"},                # besoin de voir email/id pour répondre
    "analyst": {"sales", "reviews"},                     # lecture seule agrégats, pas de PII
}


class AccessDeniedError(Exception):
    pass


def check_permission(role, resource):
    """Vérifie si un rôle a le droit d'accéder à une ressource donnée."""
    allowed = ROLE_PERMISSIONS.get(role, set())
    return resource in allowed


def require_permission(resource):
    """
    Décorateur RBAC : à apposer sur toute fonction d'accès aux données.
    La fonction décorée doit recevoir un paramètre 'role' en premier argument.
    """
    def decorator(func):
        @wraps(func)
        def wrapper(role, *args, **kwargs):
            if not check_permission(role, resource):
                raise AccessDeniedError(
                    f"Rôle '{role}' non autorisé à accéder à la ressource '{resource}'"
                )
            return func(role, *args, **kwargs)
        return wrapper
    return decorator


def list_permissions(role):
    """Retourne la liste des ressources accessibles pour un rôle donné."""
    return sorted(ROLE_PERMISSIONS.get(role, set()))
```

```python
%%writefile /content/projet_ecommerce/src/security/audit_logger.py
# ==========================================================
# src/security/audit_logger.py
# Audit logs — qui accède à quoi et quand (Phase 4E)
# ==========================================================

import os
import json
import hashlib
from datetime import datetime, timezone

AUDIT_LOG_PATH = "/content/projet_ecommerce/data/security/audit_log.jsonl"


def _mask_sensitive(value):
    """Masque toute valeur ressemblant à un email/ID avant écriture en log (exigence cahier des charges)."""
    if value is None:
        return None
    s = str(value)
    if "@" in s:
        local, domain = s.split("@", 1)
        return f"{local[0]}***@{domain}"
    if len(s) > 4:
        return s[:2] + "***" + s[-2:]
    return "***"


def log_access(role, resource, action, target_id=None, status="success"):
    """
    Enregistre un accès dans le journal d'audit.
    target_id est automatiquement masqué (aucun email/ID en clair dans les logs).
    """
    os.makedirs(os.path.dirname(AUDIT_LOG_PATH), exist_ok=True)

    entry = {
        "timestamp_utc": datetime.now(timezone.utc).isoformat(),
        "role": role,
        "resource": resource,
        "action": action,
        "target_id_masked": _mask_sensitive(target_id),
        "status": status,
    }

    with open(AUDIT_LOG_PATH, "a", encoding="utf-8") as f:
        f.write(json.dumps(entry, ensure_ascii=False) + "\n")

    return entry


def read_audit_log(n_last=20):
    """Lit les n dernières entrées du journal d'audit."""
    if not os.path.exists(AUDIT_LOG_PATH):
        return []
    with open(AUDIT_LOG_PATH, "r", encoding="utf-8") as f:
        lines = f.readlines()
    return [json.loads(l) for l in lines[-n_last:]]
```

```python
%%writefile /content/projet_ecommerce/src/security/retention_manager.py
# ==========================================================
# src/security/retention_manager.py
# Politique de rétention automatique — Phase 4E
# Basée sur les durées définies dans registre_traitements.yaml
# ==========================================================

import pandas as pd
from datetime import datetime, timezone

# Durées en jours, cohérentes avec le registre des traitements (TR-001, TR-002, TR-003)
RETENTION_POLICIES = {
    "sales": 365 * 3,     # 3 ans (obligation comptable)
    "reviews": 365 * 5,   # 5 ans
    "tickets": 365 * 2,   # 2 ans après clôture
}


def find_expired_records(df, date_column, source, reference_date=None):
    """
    Identifie les enregistrements dépassant la durée de rétention autorisée.
    Ne supprime rien automatiquement -- retourne la liste pour validation/audit
    avant toute purge réelle (bonne pratique : pas de suppression silencieuse).
    """
    reference_date = reference_date or datetime.now(timezone.utc).replace(tzinfo=None)
    retention_days = RETENTION_POLICIES.get(source)
    if retention_days is None:
        raise ValueError(f"Aucune politique de rétention définie pour '{source}'")

    cutoff = reference_date - pd.Timedelta(days=retention_days)
    expired_mask = df[date_column] < cutoff
    n_expired = expired_mask.sum()

    print(f"[find_expired_records] '{source}': {n_expired} enregistrements dépassent "
          f"{retention_days} jours de rétention (cutoff={cutoff.date()})")

    return df[expired_mask], n_expired
```

```python
%%writefile /content/projet_ecommerce/src/security/encryption.py
# ==========================================================
# src/security/encryption.py
# Chiffrement au repos des fichiers sensibles — Phase 4E
# ==========================================================

import os
from cryptography.fernet import Fernet


def encrypt_file(filepath, fernet, output_path=None):
    """Chiffre un fichier entier (ex: export JSON contenant des PII)."""
    output_path = output_path or f"{filepath}.enc"
    with open(filepath, "rb") as f:
        data = f.read()
    encrypted = fernet.encrypt(data)
    with open(output_path, "wb") as f:
        f.write(encrypted)
    print(f"[encrypt_file] {filepath} -> {output_path} ({len(encrypted)} octets chiffrés)")
    return output_path


def decrypt_file(encrypted_path, fernet, output_path=None):
    """Déchiffre un fichier préalablement chiffré avec encrypt_file."""
    output_path = output_path or encrypted_path.replace(".enc", "")
    with open(encrypted_path, "rb") as f:
        encrypted_data = f.read()
    decrypted = fernet.decrypt(encrypted_data)
    with open(output_path, "wb") as f:
        f.write(decrypted)
    print(f"[decrypt_file] {encrypted_path} -> {output_path}")
    return output_path
```

```python
%%writefile /content/projet_ecommerce/src/visualization/insight.py
# ==========================================================
# src/visualization/insight.py
# Livrable Phase 5 — Analyse & Visualisation
# Fonctions matplotlib/seaborn, exécutables inline en notebook
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud

sns.set_style("whitegrid")


# ==========================================================
# BLOC 1 — VISION GLOBALE
# ==========================================================

def show_kpis(fact_sales, fact_reviews, fact_tickets):
    ca = fact_sales["revenue"].sum()
    n_orders = len(fact_sales)
    avg_rating = fact_reviews["rating"].mean()
    n_tickets = len(fact_tickets)

    print("="*60)
    print("KPIs GLOBAUX")
    print("="*60)
    print(f"Chiffre d'affaires    : {ca:,.2f} €")
    print(f"Nombre de commandes   : {n_orders:,}")
    print(f"Note moyenne          : {avg_rating:.2f} / 5")
    print(f"Nombre de tickets     : {n_tickets:,}")


def plot_temporal_evolution(fact_sales, fact_reviews):
    sales_m = fact_sales.copy()
    sales_m["month"] = sales_m["transaction_date_clean"].dt.to_period("M").astype(str)
    monthly_revenue = sales_m.groupby("month")["revenue"].sum()

    reviews_m = fact_reviews.copy()
    reviews_m["month"] = reviews_m["review_date_clean"].dt.to_period("M").astype(str)
    monthly_reviews = reviews_m.groupby("month").size()

    fig, axes = plt.subplots(2, 1, figsize=(14, 8))
    monthly_revenue.plot(ax=axes[0], marker="o", color="steelblue")
    axes[0].set_title("CA mensuel")
    axes[0].tick_params(axis="x", rotation=45)

    monthly_reviews.plot(ax=axes[1], marker="o", color="darkorange")
    axes[1].set_title("Nombre d'avis mensuel")
    axes[1].tick_params(axis="x", rotation=45)

    plt.tight_layout()
    plt.show()


def plot_geo_distribution(dim_customers):
    plt.figure(figsize=(10, 5))
    dim_customers["country"].value_counts().plot(kind="bar", color="teal")
    plt.title("Répartition géographique des clients")
    plt.ylabel("Nombre de clients")
    plt.show()


# ==========================================================
# BLOC 2 — ANALYSE DES AVIS
# ==========================================================

def plot_rating_before_after(reviews_silver):
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    reviews_silver["rating_original"].astype(str).value_counts().head(15).plot(
        kind="bar", ax=axes[0], color="salmon")
    axes[0].set_title("Notes AVANT nettoyage (brut)")
    axes[0].tick_params(axis="x", rotation=45)

    reviews_silver["rating"].value_counts().sort_index().plot(
        kind="bar", ax=axes[1], color="seagreen")
    axes[1].set_title("Notes APRÈS nettoyage (1-5)")

    plt.tight_layout()
    plt.show()


def show_top_products(product_features, n=10):
    print("Top produits les mieux notés :")
    display(product_features.sort_values("sentiment_score", ascending=False)
            [["product_id", "avg_rating", "n_reviews"]].head(n))

    print("\nTop produits les moins bien notés :")
    display(product_features.sort_values("sentiment_score", ascending=True)
            [["product_id", "avg_rating", "n_reviews"]].head(n))


def plot_wordclouds(fact_reviews):
    positive = fact_reviews[fact_reviews["rating"] >= 4]["review_text"].dropna().astype(str)
    negative = fact_reviews[fact_reviews["rating"] <= 2]["review_text"].dropna().astype(str)

    pos_text = " ".join(positive.sample(min(3000, len(positive)), random_state=42))
    neg_text = " ".join(negative.sample(min(3000, len(negative)), random_state=42))

    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    wc_pos = WordCloud(width=600, height=400, background_color="white", colormap="Greens").generate(pos_text)
    axes[0].imshow(wc_pos); axes[0].axis("off"); axes[0].set_title("Avis positifs")

    wc_neg = WordCloud(width=600, height=400, background_color="white", colormap="Reds").generate(neg_text)
    axes[1].imshow(wc_neg); axes[1].axis("off"); axes[1].set_title("Avis négatifs")

    plt.tight_layout()
    plt.show()


def show_suspect_reviews(fact_reviews, n=20):
    """Heuristique provisoire (compte récent + note max) -- sera remplacée par le modèle Phase 6."""
    suspect = fact_reviews[
        (fact_reviews["metadata_account_age_days"] < 30) &
        (fact_reviews["rating"] == 5)
    ]
    print(f"{len(suspect)} avis potentiellement suspects (compte < 30 jours + note 5)")
    display(suspect[["review_id", "product_id", "rating", "metadata_account_age_days"]].head(n))


# ==========================================================
# BLOC 3 — SUPPORT CLIENT
# ==========================================================

def classify_ticket_type(message):
    """Classification heuristique provisoire par mots-clés (remplacée par NLP en Phase 6)."""
    m = str(message).lower()
    if any(w in m for w in ["remboursement", "rembours", "argent"]):
        return "Remboursement"
    if any(w in m for w in ["cassé", "endommagé", "damaged", "broken", "défectueux", "defectueux"]):
        return "Retour produit défectueux"
    if any(w in m for w in ["reçu", "recue", "arrivé", "arrive", "livraison", "delai", "délai"]):
        return "Problème de livraison"
    if any(w in m for w in ["facture", "montant", "erreur"]):
        return "Facturation"
    return "Question générale"


def plot_ticket_types(fact_tickets):
    tickets_typed = fact_tickets.copy()
    tickets_typed["ticket_type"] = tickets_typed["message"].apply(classify_ticket_type)

    plt.figure(figsize=(8, 8))
    tickets_typed["ticket_type"].value_counts().plot(kind="pie", autopct="%1.1f%%")
    plt.title("Répartition des tickets par type (heuristique)")
    plt.ylabel("")
    plt.show()

    return tickets_typed


def plot_delay_distribution(fact_tickets):
    avg_delay = fact_tickets["days_since_purchase"].mean()
    print(f"Délai moyen achat -> ticket : {avg_delay:.1f} jours")

    plt.figure(figsize=(10, 5))
    fact_tickets["days_since_purchase"].dropna().plot(kind="hist", bins=30, color="mediumpurple")
    plt.title("Distribution du délai achat -> ticket")
    plt.xlabel("Jours")
    plt.show()


def plot_ticket_sentiment_correlation(product_features):
    plt.figure(figsize=(8, 6))
    plt.scatter(product_features["ticket_ratio"], product_features["sentiment_score"], alpha=0.5)
    plt.xlabel("Ratio tickets/ventes")
    plt.ylabel("Score de sentiment")
    plt.title("Corrélation tickets ↔ notes produits")
    plt.show()

    corr = product_features[["ticket_ratio", "sentiment_score"]].corr().iloc[0, 1]
    print(f"Coefficient de corrélation : {corr:.3f}")


# ==========================================================
# BLOC 5 — RGPD & SÉCURITÉ
# ==========================================================

def show_rgpd_security_summary(audit_log_path, registre_path):
    import json
    import yaml
    import os

    print("="*60)
    print("CONFORMITÉ RGPD & SÉCURITÉ")
    print("="*60)

    if os.path.exists(audit_log_path):
        with open(audit_log_path, "r", encoding="utf-8") as f:
            entries = [json.loads(l) for l in f.readlines()]
        print(f"Accès journalisés : {len(entries)}")
        display(pd.DataFrame(entries).tail(10))
    else:
        print("Aucun log d'audit trouvé.")

    if os.path.exists(registre_path):
        with open(registre_path, "r", encoding="utf-8") as f:
            registre = yaml.safe_load(f)
        print(f"\nTraitements documentés : {len(registre['traitements'])}")
        for t in registre["traitements"]:
            print(f"  - {t['id']}: {t['nom']} ({t['base_legale']})")
```

```python
%%writefile /content/projet_ecommerce/src/transform/build_dimensions.py
# ==========================================================
# src/transform/build_dimensions.py
# Construction des dimensions du schéma en étoile
# ==========================================================

import pandas as pd


def build_dim_customers(df_sales, df_reviews, df_tickets):
    """
    Dimension client : union des customer_id vus dans les 3 sources.
    Le pays est complété en 2 passes : sales en priorité (source la plus fiable),
    puis metadata_country des reviews pour les clients absents de sales.
    """
    cust_sales = df_sales[["customer_id", "country"]].drop_duplicates("customer_id")

    all_ids = pd.concat([
        df_sales["customer_id"],
        df_reviews["user_id"].rename("customer_id"),
        df_tickets["customer_id"],
    ]).dropna().unique()

    dim_customers = pd.DataFrame({"customer_id": all_ids})
    dim_customers = dim_customers.merge(cust_sales, on="customer_id", how="left")

    # Complément via metadata_country des reviews pour les clients sans historique d'achat
    if "metadata_country" in df_reviews.columns:
        cust_reviews_country = (
            df_reviews[["user_id", "metadata_country"]]
            .dropna(subset=["metadata_country"])
            .drop_duplicates("user_id")
            .rename(columns={"user_id": "customer_id", "metadata_country": "country_from_reviews"})
        )
        dim_customers = dim_customers.merge(cust_reviews_country, on="customer_id", how="left")
        dim_customers["country"] = dim_customers["country"].fillna(dim_customers["country_from_reviews"])
        dim_customers = dim_customers.drop(columns=["country_from_reviews"])

    n_unknown_before = dim_customers["country"].isna().sum()
    dim_customers["country"] = dim_customers["country"].fillna("UNKNOWN")

    print(f"[dim_customers] {len(dim_customers)} clients uniques (sales+reviews+tickets)")
    print(f"[dim_customers] {n_unknown_before} clients restent sans pays connu "
          f"({n_unknown_before/len(dim_customers)*100:.1f}%) -- absents de sales ET sans metadata_country")
    return dim_customers


def build_dim_products(df_sales, df_reviews):
    all_products = pd.concat([
        df_sales["product_id"],
        df_reviews["product_id"],
    ]).dropna().unique()

    dim_products = pd.DataFrame({"product_id": sorted(all_products)})

    print(f"[dim_products] {len(dim_products)} produits uniques (sales+reviews)")
    return dim_products


def build_dim_time(df_sales, df_reviews, df_tickets):
    all_dates = pd.concat([
        df_sales["transaction_date_clean"],
        df_reviews["review_date_clean"],
        df_tickets["date_clean"],
    ]).dropna()

    date_min, date_max = all_dates.min().normalize(), all_dates.max().normalize()
    dim_time = pd.DataFrame({"date": pd.date_range(date_min, date_max, freq="D")})
    dim_time["year"] = dim_time["date"].dt.year
    dim_time["month"] = dim_time["date"].dt.month
    dim_time["day"] = dim_time["date"].dt.day
    dim_time["day_of_week"] = dim_time["date"].dt.dayofweek
    dim_time["is_weekend"] = dim_time["day_of_week"].isin([5, 6])

    print(f"[dim_time] {len(dim_time)} jours ({date_min.date()} -> {date_max.date()})")
    return dim_time
```

```python
%%writefile /content/projet_ecommerce/src/models/fraud_labeling.py
# ==========================================================
# src/models/fraud_labeling.py
# Construction du label fraude (Phase 6, Tâche A)
# Heuristique : rythme de publication (reviews/jour)
# ==========================================================

import pandas as pd
import numpy as np


def compute_review_rate(df_reviews):
    """
    Calcule le rythme de publication par avis : total_reviews / account_age_days.
    account_age_days=0 est traité comme 1 jour (évite division par zéro,
    et un compte créé le jour même qui poste plusieurs avis est déjà suspect).
    """
    df = df_reviews.copy()
    safe_age = df["metadata_account_age_days"].fillna(1).clip(lower=1)
    df["reviews_per_day"] = df["metadata_total_reviews"].fillna(0) / safe_age
    return df


def build_fraud_label(df_reviews, threshold=None, percentile=99):
    """
    Construit le label binaire fraude/légitime.
    Si threshold n'est pas fourni, il est calculé automatiquement au percentile
    spécifié de la distribution (par défaut: 99e percentile = 1% des avis les
    plus suspects, cohérent avec la contrainte du cahier des charges
    "classe minoritaire prioritaire").
    """
    df = compute_review_rate(df_reviews)

    if threshold is None:
        threshold = df["reviews_per_day"].quantile(percentile / 100)

    df["is_fraud"] = (df["reviews_per_day"] > threshold).astype(int)

    n_fraud = df["is_fraud"].sum()
    print(f"[build_fraud_label] Seuil retenu : {threshold:.3f} reviews/jour")
    print(f"[build_fraud_label] {n_fraud} avis labellisés frauduleux ({n_fraud/len(df)*100:.2f}%)")
    print(f"[build_fraud_label] {len(df) - n_fraud} avis labellisés légitimes")

    return df, threshold
```

```python
%%writefile /content/projet_ecommerce/src/models/fraud_labeling.py
# ==========================================================
# src/models/fraud_labeling.py
# Construction du label fraude (Phase 6, Tâche A)
# Heuristique : rythme de publication (reviews/jour), seuil P97
# ==========================================================

import pandas as pd
import numpy as np


def compute_review_rate(df_reviews):
    df = df_reviews.copy()
    safe_age = df["metadata_account_age_days"].fillna(1).clip(lower=1)
    df["reviews_per_day"] = df["metadata_total_reviews"].fillna(0) / safe_age
    return df


def build_fraud_label(df_reviews, threshold=None, percentile=97):
    """
    Label binaire fraude/légitime basé sur le rythme de publication.
    threshold calculé automatiquement au percentile spécifié si non fourni.
    """
    df = compute_review_rate(df_reviews)

    if threshold is None:
        threshold = df["reviews_per_day"].quantile(percentile / 100)

    df["is_fraud"] = (df["reviews_per_day"] > threshold).astype(int)

    n_fraud = df["is_fraud"].sum()
    print(f"[build_fraud_label] Seuil retenu : {threshold:.3f} reviews/jour (P{percentile})")
    print(f"[build_fraud_label] {n_fraud} avis labellisés frauduleux ({n_fraud/len(df)*100:.2f}%)")
    print(f"[build_fraud_label] {len(df) - n_fraud} avis labellisés légitimes")

    return df, threshold
```

```python
%%writefile /content/projet_ecommerce/src/models/fraud_features.py
# ==========================================================
# src/models/fraud_features.py
# Feature engineering pour la détection de fraude (Phase 6, Tâche A)
#
# EXCLUSION VOLONTAIRE : account_age_days et total_reviews sont exclus
# des features car ils ont servi à CONSTRUIRE le label (fuite de label).
# Les modèles doivent détecter la fraude via des signaux indépendants :
# caractéristiques textuelles, helpful_votes, rating, plateforme.
# ==========================================================

import re
import pandas as pd
import numpy as np

EMOJI_PATTERN = re.compile(
    "["
    "\U0001F300-\U0001FAFF"
    "\U00002600-\U000027BF"
    "]+", flags=re.UNICODE
)
LINK_PATTERN = re.compile(r"https?://|www\.", re.IGNORECASE)


def build_text_features(df):
    """Caractéristiques textuelles du review_text (cahier des charges Tâche A)."""
    text = df["review_text"].fillna("").astype(str)

    df["text_length"] = text.str.len()
    df["word_count"] = text.str.split().str.len()
    df["has_link"] = text.str.contains(LINK_PATTERN).astype(int)
    df["has_emoji"] = text.apply(lambda t: bool(EMOJI_PATTERN.search(t))).astype(int)
    df["exclamation_count"] = text.str.count("!")
    df["uppercase_ratio"] = text.apply(
        lambda t: sum(1 for c in t if c.isupper()) / len(t) if len(t) > 0 else 0
    )

    # Répétition : nombre de fois où EXACTEMENT ce texte apparaît dans tout le corpus
    text_counts = text.value_counts()
    df["duplicate_text_count"] = text.map(text_counts)

    return df


def build_fraud_feature_set(df_labeled):
    """
    Assemble le jeu de features final pour l'entraînement.
    Colonnes explicitement exclues : metadata_account_age_days, metadata_total_reviews
    (utilisées uniquement pour construire le label, jamais comme feature).
    """
    df = build_text_features(df_labeled.copy())

    # One-hot encoding de la plateforme
    platform_dummies = pd.get_dummies(df["metadata_platform"], prefix="platform", dummy_na=True)

    feature_columns = [
        "rating", "metadata_helpful_votes",
        "text_length", "word_count", "has_link", "has_emoji",
        "exclamation_count", "uppercase_ratio", "duplicate_text_count",
    ]

    X = pd.concat([df[feature_columns], platform_dummies], axis=1)
    X["metadata_helpful_votes"] = X["metadata_helpful_votes"].fillna(0)
    y = df["is_fraud"]

    print(f"[build_fraud_feature_set] {X.shape[1]} features, {len(X)} observations")
    print(f"[build_fraud_feature_set] Features utilisées : {list(X.columns)}")

    return X, y
```

```python
%%writefile /content/projet_ecommerce/src/models/fraud_split.py
# ==========================================================
# src/models/fraud_split.py
# Split train/test stratifié + gestion du déséquilibre (Phase 6, Tâche A)
# ==========================================================

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


def prepare_train_test(X, y, test_size=0.2, random_state=42, drop_constant=True):
    """
    Split stratifié (préserve le ratio 3%/97% dans train ET test).
    Retire les colonnes à variance nulle si demandé.
    Standardise les features numériques (utile pour le réseau dense).
    """
    X = X.copy()

    if drop_constant:
        constant_cols = [c for c in X.columns if X[c].nunique() <= 1]
        if constant_cols:
            print(f"[prepare_train_test] Colonnes constantes retirées : {constant_cols}")
            X = X.drop(columns=constant_cols)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=random_state, stratify=y
    )

    scaler = StandardScaler()
    numeric_cols = X_train.select_dtypes(include=["int64", "float64"]).columns
    X_train_scaled = X_train.copy()
    X_test_scaled = X_test.copy()
    X_train_scaled[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
    X_test_scaled[numeric_cols] = scaler.transform(X_test[numeric_cols])

    print(f"[prepare_train_test] Train : {X_train.shape}, ratio fraude = {y_train.mean()*100:.2f}%")
    print(f"[prepare_train_test] Test  : {X_test.shape}, ratio fraude = {y_test.mean()*100:.2f}%")

    return X_train, X_test, y_train, y_test, X_train_scaled, X_test_scaled, scaler
```

# Étape suivante : entraînement des 3 algorithmes classiques

RandomForest, XGBoost, LightGBM, avec gestion du déséquilibre via pondération de classe (pas de sur-échantillonnage synthétique, pour rester fidèle à la distribution réelle), et les métriques Precision, Recall, F1, AUC-ROC en priorisant la classe minoritaire.

```python
%%writefile /content/projet_ecommerce/src/models/train_classical.py
# ==========================================================
# src/models/train_classical.py
# Entraînement RandomForest, XGBoost, LightGBM (Phase 6, Tâche A)
# Gestion du déséquilibre via pondération de classe (pas de SMOTE)
# ==========================================================

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    precision_score, recall_score, f1_score, roc_auc_score,
    confusion_matrix, classification_report
)
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


def evaluate_model(model, X_test, y_test, model_name):
    """Calcule les métriques prioritaires pour la classe minoritaire (fraude=1)."""
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    metrics = {
        "model": model_name,
        "precision": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1),
        "f1": f1_score(y_test, y_pred, pos_label=1),
        "auc_roc": roc_auc_score(y_test, y_proba),
    }

    print(f"\n=== {model_name} ===")
    print(f"Precision (fraude) : {metrics['precision']:.3f}")
    print(f"Recall (fraude)    : {metrics['recall']:.3f}")
    print(f"F1-score (fraude)  : {metrics['f1']:.3f}")
    print(f"AUC-ROC            : {metrics['auc_roc']:.3f}")
    print("\nMatrice de confusion :")
    print(confusion_matrix(y_test, y_pred))
    print("\nRapport complet :")
    print(classification_report(y_test, y_pred, target_names=["légitime", "fraude"]))

    return metrics, y_pred, y_proba


def train_random_forest(X_train, y_train, X_test, y_test):
    """class_weight='balanced' : pondère automatiquement la classe minoritaire."""
    model = RandomForestClassifier(
        n_estimators=200, max_depth=10, class_weight="balanced",
        random_state=42, n_jobs=-1
    )
    model.fit(X_train, y_train)
    metrics, y_pred, y_proba = evaluate_model(model, X_test, y_test, "RandomForest")
    return model, metrics, y_pred, y_proba


def train_xgboost(X_train, y_train, X_test, y_test):
    """scale_pos_weight = ratio négatifs/positifs, équivalent XGBoost du class_weight."""
    scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()
    model = XGBClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.1,
        scale_pos_weight=scale_pos_weight,
        eval_metric="logloss", random_state=42, n_jobs=-1
    )
    model.fit(X_train, y_train)
    metrics, y_pred, y_proba = evaluate_model(model, X_test, y_test, "XGBoost")
    return model, metrics, y_pred, y_proba


def train_lightgbm(X_train, y_train, X_test, y_test):
    model = LGBMClassifier(
        n_estimators=200, max_depth=6, learning_rate=0.1,
        class_weight="balanced", random_state=42, n_jobs=-1, verbose=-1
    )
    model.fit(X_train, y_train)
    metrics, y_pred, y_proba = evaluate_model(model, X_test, y_test, "LightGBM")
    return model, metrics, y_pred, y_proba
```

# Étape suivante : réseau de neurones dense (Keras)

```python
%%writefile /content/projet_ecommerce/src/models/train_dense_nn.py
# ==========================================================
# src/models/train_dense_nn.py
# Réseau de neurones dense (Phase 6, Tâche A)
# Gestion du déséquilibre via class_weight (cohérent avec les 3 autres modèles)
# ==========================================================

import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import (
    precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
)


def build_dense_model(input_dim):
    model = keras.Sequential([
        keras.layers.Input(shape=(input_dim,)),
        keras.layers.Dense(64, activation="relu"),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(32, activation="relu"),
        keras.layers.Dropout(0.3),
        keras.layers.Dense(16, activation="relu"),
        keras.layers.Dense(1, activation="sigmoid"),
    ])
    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=[keras.metrics.AUC(name="auc"), keras.metrics.Recall(name="recall")]
    )
    return model


def train_dense_nn(X_train_scaled, y_train, X_test_scaled, y_test, epochs=30, batch_size=256):
    """
    Utilise les features standardisées (X_*_scaled), essentiel pour un réseau dense.
    class_weight rééquilibre la fonction de perte, cohérent avec class_weight='balanced'
    des modèles classiques.
    """
    input_dim = X_train_scaled.shape[1]
    model = build_dense_model(input_dim)

    n_neg = (y_train == 0).sum()
    n_pos = (y_train == 1).sum()
    class_weight = {0: 1.0, 1: n_neg / n_pos}
    print(f"[train_dense_nn] class_weight : {class_weight}")

    early_stop = keras.callbacks.EarlyStopping(
        monitor="val_auc", mode="max", patience=5, restore_best_weights=True
    )

    history = model.fit(
        X_train_scaled.values.astype("float32"), y_train.values.astype("float32"),
        validation_split=0.2,
        epochs=epochs, batch_size=batch_size,
        class_weight=class_weight,
        callbacks=[early_stop],
        verbose=1
    )

    y_proba = model.predict(X_test_scaled.values.astype("float32")).flatten()
    y_pred = (y_proba > 0.5).astype(int)

    metrics = {
        "model": "DenseNN",
        "precision": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1),
        "f1": f1_score(y_test, y_pred, pos_label=1),
        "auc_roc": roc_auc_score(y_test, y_proba),
    }

    print(f"\n=== DenseNN ===")
    print(f"Precision (fraude) : {metrics['precision']:.3f}")
    print(f"Recall (fraude)    : {metrics['recall']:.3f}")
    print(f"F1-score (fraude)  : {metrics['f1']:.3f}")
    print(f"AUC-ROC            : {metrics['auc_roc']:.3f}")
    print("\nMatrice de confusion :")
    print(confusion_matrix(y_test, y_pred))
    print("\nRapport complet :")
    print(classification_report(y_test, y_pred, target_names=["légitime", "fraude"]))

    return model, metrics, y_pred, y_proba, history
```

# Explicabilité SHAP (obligatoire, Art. 22 RGPD)

```python
%%writefile /content/projet_ecommerce/src/models/fraud_explainability.py
# ==========================================================
# src/models/fraud_explainability.py
# Explicabilité SHAP (Art. 22 RGPD) — Phase 6, Tâche A
# ==========================================================

import shap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def compute_shap_values(model, X_test, sample_size=2000, random_state=42):
    """
    Calcule les valeurs SHAP sur un échantillon (TreeExplainer est rapide,
    mais on limite quand même la taille pour la lisibilité des graphiques).
    """
    X_sample = X_test.sample(min(sample_size, len(X_test)), random_state=random_state)
    explainer = shap.TreeExplainer(model)
    shap_values = explainer.shap_values(X_sample)

    # RandomForest binaire : shap_values peut être une liste [classe0, classe1] ou un array 3D
    if isinstance(shap_values, list):
        shap_values_fraud = shap_values[1]
    elif shap_values.ndim == 3:
        shap_values_fraud = shap_values[:, :, 1]
    else:
        shap_values_fraud = shap_values

    return explainer, shap_values_fraud, X_sample


def plot_shap_summary(shap_values_fraud, X_sample):
    plt.figure(figsize=(10, 6))
    shap.summary_plot(shap_values_fraud, X_sample, show=False)
    plt.title("Importance des features (SHAP) — classe fraude")
    plt.tight_layout()
    plt.show()


def get_feature_importance_ranking(shap_values_fraud, X_sample):
    """Classement des features par importance moyenne absolue (interprétation globale)."""
    importance = np.abs(shap_values_fraud).mean(axis=0)
    ranking = pd.DataFrame({
        "feature": X_sample.columns,
        "mean_abs_shap": importance
    }).sort_values("mean_abs_shap", ascending=False)
    return ranking


def explain_single_prediction(explainer, model, X_test, y_test, index):
    """
    Explication individuelle d'une prédiction (exigence Art. 22 :
    droit à une explication pour une décision automatisée individuelle).
    """
    instance = X_test.loc[[index]]
    shap_val = explainer.shap_values(instance)
    if isinstance(shap_val, list):
        shap_val = shap_val[1]
    elif shap_val.ndim == 3:
        shap_val = shap_val[:, :, 1]

    proba = model.predict_proba(instance)[0, 1]
    actual = y_test.loc[index]

    print(f"Avis index={index} | Probabilité de fraude prédite : {proba:.3f} | Label réel : {actual}")
    print("\nContribution de chaque feature à cette prédiction :")
    contrib = pd.DataFrame({
        "feature": instance.columns,
        "value": instance.values[0],
        "shap_contribution": shap_val[0]
    }).sort_values("shap_contribution", key=abs, ascending=False)
    return contrib
```

```python
%%writefile /content/projet_ecommerce/src/models/fraud_labeling.py
# ==========================================================
# src/models/fraud_labeling.py
# Construction du label fraude (Phase 6, Tâche A)
# Heuristique : rythme de publication (reviews/jour), seuil P97
#
# CORRECTIF : les avis sans metadata (29 971 lignes) sont exclus de
# l'entraînement. Leur account_age/total_reviews étant imputés par
# défaut (fillna 1/0), leur reviews_per_day vaut toujours 0, ce qui
# les labellise systématiquement "légitime" -- pas une vraie observation,
# mais un artefact de l'imputation (fuite de label indirecte détectée
# via SHAP : platform_nan devenait la feature la plus importante).
# ==========================================================

import pandas as pd
import numpy as np


def compute_review_rate(df_reviews):
    df = df_reviews.copy()
    safe_age = df["metadata_account_age_days"].fillna(1).clip(lower=1)
    df["reviews_per_day"] = df["metadata_total_reviews"].fillna(0) / safe_age
    return df


def build_fraud_label(df_reviews, threshold=None, percentile=97, exclude_missing_metadata=True):
    df = compute_review_rate(df_reviews)

    if exclude_missing_metadata:
        n_before = len(df)
        df = df[df["metadata_platform"].notna()].copy()
        n_removed = n_before - len(df)
        print(f"[build_fraud_label] {n_removed} avis sans metadata exclus "
              f"(label non fiable, artefact d'imputation)")

    if threshold is None:
        threshold = df["reviews_per_day"].quantile(percentile / 100)

    df["is_fraud"] = (df["reviews_per_day"] > threshold).astype(int)

    n_fraud = df["is_fraud"].sum()
    print(f"[build_fraud_label] Seuil retenu : {threshold:.3f} reviews/jour (P{percentile})")
    print(f"[build_fraud_label] {n_fraud} avis labellisés frauduleux ({n_fraud/len(df)*100:.2f}%)")
    print(f"[build_fraud_label] {len(df) - n_fraud} avis labellisés légitimes")
    print(f"[build_fraud_label] Total avis conservés pour modélisation : {len(df)}")

    return df, threshold
```

```python
%%writefile /content/projet_ecommerce/docs/model_cards/fraud_detection.md
# Fiche Modèle RGPD — Détection d'avis frauduleux

## 1. Identification

- **Nom du modèle** : Détection d'avis frauduleux (Phase 6, Tâche A)
- **Version** : 1.0
- **Date d'entraînement** : 2026-07-06
- **Modèles comparés** : RandomForest, XGBoost, LightGBM, Réseau dense (Keras)
- **Modèle retenu** : LightGBM (meilleur F1=0.059, AUC=0.506)

## 2. Finalité (Art. 5.1.b RGPD)

Identifier les avis clients potentiellement générés par des comptes à comportement
automatisé (bots), afin d'appuyer une revue humaine — **ce modèle n'automatise
aucune décision seul** (pas de suppression automatique d'avis).

## 3. Construction du label (méthodologie et limites)

Le label `is_fraud` est une **heuristique**, pas une vérité terrain observée :
il est défini par le rythme de publication d'un compte
(`total_reviews / account_age_days`), seuillé au 97e percentile de la
distribution (seuil = 2.705 avis/jour → ~3% des avis labellisés "fraude").

**Limite reconnue** : cette heuristique capture un seul type de comportement
suspect (publication anormalement rapide). Elle ne capture pas d'autres formes
de fraude (avis achetés, coordination entre comptes distincts, etc.).

## 4. Features utilisées — Minimisation (Art. 5.1.c RGPD)

| Feature | Justification |
|---|---|
| `rating` | Signal comportemental disponible sans donnée personnelle |
| `metadata_helpful_votes` | Engagement communautaire, non-personnel |
| `text_length`, `word_count` | Caractéristique structurelle du texte |
| `has_emoji`, `exclamation_count`, `uppercase_ratio` | Style d'écriture |
| `duplicate_text_count` | Répétition de contenu dans le corpus global |
| `platform_*` (one-hot) | Canal technique de publication |

**Features explicitement exclues** :
- `metadata_account_age_days`, `metadata_total_reviews` : utilisées pour
  construire le label, exclues des features pour éviter la fuite de label directe.
- Aucun email, nom, ou identifiant direct n'entre dans le modèle.
- Les 29 971 avis sans `metadata` (donc sans label fiable) sont exclus de
  l'entraînement (voir section 6, incident de fuite indirecte).

## 5. Performances (test set, dataset assaini)

| Modèle | Precision | Recall | F1 | AUC-ROC |
|---|---|---|---|---|
| RandomForest | 0.029 | 0.145 | 0.048 | 0.490 |
| XGBoost | 0.028 | 0.257 | 0.051 | 0.484 |
| **LightGBM** | **0.032** | **0.324** | **0.059** | **0.506** |
| Réseau dense | 0.030 | 0.826 | 0.058 | 0.510 |

**Interprétation** : AUC-ROC ≈ 0.5 pour les 4 modèles = performance au niveau
du hasard. Ceci démontre l'absence de fuite de label résiduelle, mais signifie
aussi que les features disponibles (texte, engagement, plateforme) ne portent
pas de signal prédictif sur le rythme de publication du compte. **Ce modèle
n'est pas recommandé en l'état pour une mise en production** ; il documente
une tentative méthodologiquement rigoureuse plutôt qu'un système opérationnel.

## 6. Historique des incidents de fuite de label (traçabilité méthodologique)

1. **Fuite directe** détectée avant entraînement : `account_age_days` et
   `total_reviews` auraient permis une reconstruction triviale du label
   (ils le définissent). Exclues des features dès la conception.
2. **Fuite indirecte** détectée via SHAP après un premier entraînement :
   `platform_nan` dominait l'importance des features (0.097, ~2x la 2e
   feature). Cause : les avis sans `metadata` recevaient un label "légitime"
   automatique par artefact d'imputation (`fillna` sur account_age/total_reviews).
   Correctif : exclusion des 29 971 avis sans metadata de l'entraînement.
   Après correctif, `platform_nan` devient une colonne constante (retirée
   automatiquement) et l'importance SHAP redevient homogène (max 0.042).

## 7. Explicabilité (Art. 22 RGPD)

Méthode : SHAP TreeExplainer sur le modèle RandomForest (échantillon de 2000
observations test). Chaque prédiction individuelle peut être expliquée via
`explain_single_prediction()` (contribution de chaque feature au score final),
disponible pour toute personne souhaitant contester une décision.

## 8. Contraintes RGPD respectées

- ✅ Pas de donnée personnelle directe dans les features (Art. 5.1.c)
- ✅ Explicabilité individuelle disponible (Art. 22)
- ✅ Finalité limitée à l'aide à la décision humaine, pas d'automatisation seule
- ✅ Traçabilité méthodologique complète (biais et corrections documentés)

## 9. Recommandations pour une future itération

- Envisager l'usage du RGPD-compatible `bot_score` s'il devient disponible
  nativement (au lieu d'une heuristique reconstruite)
- Explorer des features de coordination temporelle (plusieurs avis sur le
  même produit dans une fenêtre de temps courte, depuis des comptes distincts)
- Réévaluer avec un vrai jeu d'entraînement annoté manuellement si possible
```

# Objectif : automatiser ingestion → nettoyage → transformation → ML → dashboard en un DAG unique, avec gestion d'erreurs et planification.

```python
%%writefile /content/projet_ecommerce/orchestrator/pipeline.py
# ==========================================================
# orchestrator/pipeline.py
# Orchestration du pipeline ShopEuro (Phase 7)
# Alternative à un DAG Airflow/Prefect -- séquence de tâches avec
# gestion d'erreurs, retry, logs, et point d'entrée planifiable (cron).
# ==========================================================

import os
import sys
import time
import logging
import traceback
from datetime import datetime

sys.path.append("/content/projet_ecommerce")

DATA_PATH = "/content/projet_ecommerce/data"
LOG_PATH = "/content/projet_ecommerce/data/orchestrator_logs"
CSV_FILE = "/content/sales_history_big 2(in).csv"
JSON_FILE = "/content/customer_reviews_big 1.jsonl"
TXT_FILE = "/content/support_tickets_big.txt"

os.makedirs(LOG_PATH, exist_ok=True)

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[
        logging.FileHandler(f"{LOG_PATH}/pipeline_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"),
        logging.StreamHandler(sys.stdout),
    ]
)
logger = logging.getLogger("shopeuro_pipeline")


class PipelineError(Exception):
    pass


def with_retry(func, task_name, max_retries=2, delay_seconds=5):
    """Exécute une tâche avec retry automatique et alerte en cas d'échec définitif."""
    for attempt in range(1, max_retries + 2):
        try:
            logger.info(f"[{task_name}] Tentative {attempt}/{max_retries + 1}")
            result = func()
            logger.info(f"[{task_name}] Succès")
            return result
        except Exception as e:
            logger.error(f"[{task_name}] Échec tentative {attempt} : {e}")
            logger.debug(traceback.format_exc())
            if attempt <= max_retries:
                time.sleep(delay_seconds)
            else:
                send_alert(task_name, e)
                raise PipelineError(f"Échec définitif de '{task_name}' après {max_retries + 1} tentatives") from e


def send_alert(task_name, error):
    """Point d'alerte -- en production : Slack/email/PagerDuty."""
    logger.critical(f"[ALERTE] Tâche '{task_name}' a échoué définitivement : {error}")


# ----------------------------------------------------------
# TÂCHES DU PIPELINE
# ----------------------------------------------------------

def task_ingestion():
    import json
    import pandas as pd

    df_sales = pd.read_csv(CSV_FILE, encoding="utf-8", low_memory=False)

    reviews_raw = []
    with open(JSON_FILE, "r", encoding="utf-8") as f:
        for line in f:
            try:
                reviews_raw.append(json.loads(line))
            except Exception:
                continue
    df_reviews = pd.DataFrame(reviews_raw)

    os.makedirs(f"{DATA_PATH}/raw", exist_ok=True)
    df_sales.to_parquet(f"{DATA_PATH}/raw/sales_bronze.parquet", index=False)
    df_reviews.to_parquet(f"{DATA_PATH}/raw/reviews_bronze.parquet", index=False)

    logger.info(f"Ingestion : sales={len(df_sales)}, reviews={len(df_reviews)}")
    return {"n_sales": len(df_sales), "n_reviews": len(df_reviews)}


def task_cleaning():
    import pandas as pd
    from src.cleaning.clean_dates import clean_date_column
    from src.cleaning.clean_sales import clean_sales_pipeline
    from src.cleaning.clean_reviews import clean_reviews_pipeline

    df_sales = pd.read_parquet(f"{DATA_PATH}/raw/sales_bronze.parquet")
    df_sales, _ = clean_sales_pipeline(df_sales)
    df_sales, _ = clean_date_column(df_sales, "transaction_date")
    df_sales.to_parquet(f"{DATA_PATH}/cleaned/sales_silver.parquet", index=False)

    df_reviews = pd.read_parquet(f"{DATA_PATH}/raw/reviews_bronze.parquet")
    df_reviews, _ = clean_reviews_pipeline(df_reviews)
    df_reviews, _ = clean_date_column(df_reviews, "review_date")
    df_reviews.to_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet", index=False)

    logger.info(f"Nettoyage : sales={len(df_sales)}, reviews={len(df_reviews)}")
    return {"n_sales_clean": len(df_sales), "n_reviews_clean": len(df_reviews)}


def task_transformation():
    import pandas as pd
    from src.transform.build_facts import build_fact_sales, build_fact_reviews

    df_sales = pd.read_parquet(f"{DATA_PATH}/cleaned/sales_silver.parquet")
    df_reviews = pd.read_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet")

    fact_sales = build_fact_sales(df_sales)
    fact_reviews = build_fact_reviews(df_reviews)

    os.makedirs(f"{DATA_PATH}/features", exist_ok=True)
    fact_sales.to_parquet(f"{DATA_PATH}/features/fact_sales.parquet", index=False)
    fact_reviews.to_parquet(f"{DATA_PATH}/features/fact_reviews.parquet", index=False)

    logger.info("Transformation terminée")
    return {"revenue_total": float(fact_sales["revenue"].sum())}


def task_ml_training():
    import pandas as pd
    import joblib
    from src.models.fraud_labeling import build_fraud_label
    from src.models.fraud_features import build_fraud_feature_set
    from src.models.fraud_split import prepare_train_test
    from src.models.train_classical import train_lightgbm

    df_reviews = pd.read_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet")
    df_labeled, _ = build_fraud_label(df_reviews, percentile=97, exclude_missing_metadata=True)
    X, y = build_fraud_feature_set(df_labeled)
    X_train, X_test, y_train, y_test, _, _, _ = prepare_train_test(X, y)

    model, metrics, _, _ = train_lightgbm(X_train, y_train, X_test, y_test)

    os.makedirs("/content/projet_ecommerce/models", exist_ok=True)
    joblib.dump(model, "/content/projet_ecommerce/models/fraud_lightgbm_latest.pkl")

    logger.info(f"ML : AUC={metrics['auc_roc']:.3f}, F1={metrics['f1']:.3f}")
    return metrics


def task_update_dashboard():
    logger.info("Données du dashboard rafraîchies (cache invalidé)")
    return {"status": "dashboard_ready"}


# ----------------------------------------------------------
# ORCHESTRATION PRINCIPALE
# ----------------------------------------------------------

def run_pipeline():
    start_time = datetime.now()
    logger.info("="*60)
    logger.info(f"DÉMARRAGE DU PIPELINE — {start_time.isoformat()}")
    logger.info("="*60)

    results = {}
    tasks = [
        ("ingestion", task_ingestion),
        ("cleaning", task_cleaning),
        ("transformation", task_transformation),
        ("ml_training", task_ml_training),
        ("update_dashboard", task_update_dashboard),
    ]

    try:
        for task_name, task_func in tasks:
            results[task_name] = with_retry(task_func, task_name)
    except PipelineError as e:
        logger.critical(f"PIPELINE ARRÊTÉ : {e}")
        return {"status": "failed", "error": str(e), "results": results}

    duration = (datetime.now() - start_time).total_seconds()
    logger.info(f"PIPELINE TERMINÉ AVEC SUCCÈS en {duration:.1f}s")
    return {"status": "success", "duration_seconds": duration, "results": results}


if __name__ == "__main__":
    run_pipeline()
```

```python
%%writefile /content/projet_ecommerce/orchestrator/pipeline.py
# ==========================================================
# orchestrator/pipeline.py
# Orchestration du pipeline ShopEuro (Phase 7)
# ==========================================================

import os
import sys
import time
import logging
import traceback
from datetime import datetime

sys.path.append("/content/projet_ecommerce")

DATA_PATH = "/content/projet_ecommerce/data"
LOG_PATH = "/content/projet_ecommerce/data/orchestrator_logs"
CSV_FILE = "/content/sales_history_big 2(in).csv"
JSON_FILE = "/content/customer_reviews_big 1.jsonl"
TXT_FILE = "/content/support_tickets_big.txt"

os.makedirs(LOG_PATH, exist_ok=True)

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s [%(levelname)s] %(message)s",
    handlers=[
        logging.FileHandler(f"{LOG_PATH}/pipeline_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"),
        logging.StreamHandler(sys.stdout),
    ]
)
logger = logging.getLogger("shopeuro_pipeline")


class PipelineError(Exception):
    pass


def with_retry(func, task_name, max_retries=2, delay_seconds=5):
    for attempt in range(1, max_retries + 2):
        try:
            logger.info(f"[{task_name}] Tentative {attempt}/{max_retries + 1}")
            result = func()
            logger.info(f"[{task_name}] Succès")
            return result
        except Exception as e:
            logger.error(f"[{task_name}] Échec tentative {attempt} : {e}")
            logger.debug(traceback.format_exc())
            if attempt <= max_retries:
                time.sleep(delay_seconds)
            else:
                send_alert(task_name, e)
                raise PipelineError(f"Échec définitif de '{task_name}' après {max_retries + 1} tentatives") from e


def send_alert(task_name, error):
    logger.critical(f"[ALERTE] Tâche '{task_name}' a échoué définitivement : {error}")


def _make_parquet_safe(df):
    """
    Cast les colonnes 'object' à types mixtes (ex: rating brut mélangeant
    int/str) en string pour compatibilité Parquet. N'altère AUCUNE valeur,
    change uniquement le type de stockage -- la donnée reste 'brute' au
    sens métier (aucun nettoyage), seule sa sérialisation est rendue possible.
    """
    df = df.copy()
    for col in df.select_dtypes(include="object").columns:
        types_in_col = df[col].dropna().apply(type).nunique()
        if types_in_col > 1:
            df[col] = df[col].astype(str)
    return df


# ----------------------------------------------------------
# TÂCHES DU PIPELINE
# ----------------------------------------------------------

def task_ingestion():
    import json
    import pandas as pd

    df_sales = pd.read_csv(CSV_FILE, encoding="utf-8", low_memory=False)

    reviews_raw = []
    with open(JSON_FILE, "r", encoding="utf-8") as f:
        for line in f:
            try:
                reviews_raw.append(json.loads(line))
            except Exception:
                continue
    df_reviews = pd.DataFrame(reviews_raw)

    df_sales = _make_parquet_safe(df_sales)
    df_reviews = _make_parquet_safe(df_reviews)

    os.makedirs(f"{DATA_PATH}/raw", exist_ok=True)
    df_sales.to_parquet(f"{DATA_PATH}/raw/sales_bronze.parquet", index=False)
    df_reviews.to_parquet(f"{DATA_PATH}/raw/reviews_bronze.parquet", index=False)

    logger.info(f"Ingestion : sales={len(df_sales)}, reviews={len(df_reviews)}")
    return {"n_sales": len(df_sales), "n_reviews": len(df_reviews)}


def task_cleaning():
    import pandas as pd
    from src.cleaning.clean_dates import clean_date_column
    from src.cleaning.clean_sales import clean_sales_pipeline
    from src.cleaning.clean_reviews import clean_reviews_pipeline

    df_sales = pd.read_parquet(f"{DATA_PATH}/raw/sales_bronze.parquet")
    df_sales, _ = clean_sales_pipeline(df_sales)
    df_sales, _ = clean_date_column(df_sales, "transaction_date")
    df_sales.to_parquet(f"{DATA_PATH}/cleaned/sales_silver.parquet", index=False)

    df_reviews = pd.read_parquet(f"{DATA_PATH}/raw/reviews_bronze.parquet")
    df_reviews, _ = clean_reviews_pipeline(df_reviews)
    df_reviews, _ = clean_date_column(df_reviews, "review_date")
    df_reviews.to_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet", index=False)

    logger.info(f"Nettoyage : sales={len(df_sales)}, reviews={len(df_reviews)}")
    return {"n_sales_clean": len(df_sales), "n_reviews_clean": len(df_reviews)}


def task_transformation():
    import pandas as pd
    from src.transform.build_facts import build_fact_sales, build_fact_reviews

    df_sales = pd.read_parquet(f"{DATA_PATH}/cleaned/sales_silver.parquet")
    df_reviews = pd.read_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet")

    fact_sales = build_fact_sales(df_sales)
    fact_reviews = build_fact_reviews(df_reviews)

    os.makedirs(f"{DATA_PATH}/features", exist_ok=True)
    fact_sales.to_parquet(f"{DATA_PATH}/features/fact_sales.parquet", index=False)
    fact_reviews.to_parquet(f"{DATA_PATH}/features/fact_reviews.parquet", index=False)

    logger.info("Transformation terminée")
    return {"revenue_total": float(fact_sales["revenue"].sum())}


def task_ml_training():
    import pandas as pd
    import joblib
    from src.models.fraud_labeling import build_fraud_label
    from src.models.fraud_features import build_fraud_feature_set
    from src.models.fraud_split import prepare_train_test
    from src.models.train_classical import train_lightgbm

    df_reviews = pd.read_parquet(f"{DATA_PATH}/cleaned/reviews_silver.parquet")
    df_labeled, _ = build_fraud_label(df_reviews, percentile=97, exclude_missing_metadata=True)
    X, y = build_fraud_feature_set(df_labeled)
    X_train, X_test, y_train, y_test, _, _, _ = prepare_train_test(X, y)

    model, metrics, _, _ = train_lightgbm(X_train, y_train, X_test, y_test)

    os.makedirs("/content/projet_ecommerce/models", exist_ok=True)
    joblib.dump(model, "/content/projet_ecommerce/models/fraud_lightgbm_latest.pkl")

    logger.info(f"ML : AUC={metrics['auc_roc']:.3f}, F1={metrics['f1']:.3f}")
    return metrics


def task_update_dashboard():
    logger.info("Données du dashboard rafraîchies (cache invalidé)")
    return {"status": "dashboard_ready"}


# ----------------------------------------------------------
# ORCHESTRATION PRINCIPALE
# ----------------------------------------------------------

def run_pipeline():
    start_time = datetime.now()
    logger.info("="*60)
    logger.info(f"DÉMARRAGE DU PIPELINE — {start_time.isoformat()}")
    logger.info("="*60)

    results = {}
    tasks = [
        ("ingestion", task_ingestion),
        ("cleaning", task_cleaning),
        ("transformation", task_transformation),
        ("ml_training", task_ml_training),
        ("update_dashboard", task_update_dashboard),
    ]

    try:
        for task_name, task_func in tasks:
            results[task_name] = with_retry(task_func, task_name)
    except PipelineError as e:
        logger.critical(f"PIPELINE ARRÊTÉ : {e}")
        return {"status": "failed", "error": str(e), "results": results}

    duration = (datetime.now() - start_time).total_seconds()
    logger.info(f"PIPELINE TERMINÉ AVEC SUCCÈS en {duration:.1f}s")
    return {"status": "success", "duration_seconds": duration, "results": results}


if __name__ == "__main__":
    run_pipeline()
```

```python
readme_content = '''# ShopEuro — E-commerce Intelligence

Pipeline Big Data de bout en bout pour ShopEuro : detection de fraude sur les avis,
classification d'urgence des tickets support, conformite RGPD complete, et dashboard
d'analyse -- construit a partir de 3 sources heterogenes (CSV, JSONL, TXT) representant
~225 Mo de donnees brutes et bruitees.

## Problematique

Comment concevoir un pipeline Big Data conforme au RGPD, capable de detecter les
signaux faibles d'insatisfaction client et de fraude, en combinant analyse de
sentiments et apprentissage automatique, tout en garantissant la protection des
donnees personnelles ?

## Architecture du projet

    projet_ecommerce/
    |-- data/               # Bronze (brut) / Silver (nettoye) / Gold (features)
    |-- notebooks/          # EDA et notebooks ML
    |-- src/
    |   |-- cleaning/       # Nettoyage : dates, sales, reviews, tickets
    |   |-- transform/      # Fusion, dimensions, faits, features
    |   |-- models/         # Detection de fraude (4 modeles + SHAP)
    |   |-- rgpd/           # Pseudonymisation, droit a l'oubli, portabilite
    |   |-- security/       # RBAC, audit logs, chiffrement, retention
    |   |-- visualization/  # Dashboard (insight.py + Streamlit)
    |-- app/                # Application Streamlit
    |-- orchestrator/       # Pipeline automatise
    |-- rgpd/               # Registre des traitements (YAML), certificats
    |-- tests/              # Tests de conformite
    |-- docs/               # Rapport technique, fiches modeles RGPD

## Pipeline (7 phases)

1. EDA -- exploration des 3 sources brutes, cartographie PII
2. Nettoyage -- dates multi-formats, pays/quantites/prix aberrants, HTML, parsing multi-format des tickets
3. Transformation -- schema en etoile, jointures, features derivees, Parquet Bronze/Silver/Gold
4. RGPD & Securite -- chiffrement, bruit differentiel, masquage, droit a l'oubli/portabilite, RBAC, audit
5. Dashboard -- KPIs, analyse avis/support, correlations, conformite RGPD
6. ML/DL -- detection d'avis frauduleux (RandomForest, XGBoost, LightGBM, reseau dense), SHAP
7. Orchestration -- pipeline automatise avec retry et alertes

## Installation

    pip install -r requirements.txt

## Execution

    make run       # pipeline complet
    make test      # tests de conformite RGPD
    make dashboard # lance le dashboard

## Points methodologiques notables

- Jointure tickets<->sales impossible via order_id : les identifiants ORD_xxx
  cites dans les tickets ne correspondent a aucun transaction_id reel. Jointure
  de repli via customer_id + proximite temporelle (merge_asof).
- Deux fuites de label detectees et corrigees lors de l'entrainement du modele
  de fraude (directe : account_age_days/total_reviews ; indirecte : artefact
  d'imputation sur les avis sans metadata), documentees dans la fiche modele RGPD.
- Anomalie temporelle : chute brutale du volume de donnees a partir de
  juillet 2025 (dataset annonce sur 6 mois, couvre en realite ~18 mois d'activite
  stable + 6 mois de donnees residuelles).

## Auteur

Projet realise dans le cadre du M1 Intelligence Artificielle -- SUP'RH Business
School & AI, laboratoire CERMACH.
'''

with open("/content/projet_ecommerce/README.md", "w", encoding="utf-8") as f:
    f.write(readme_content)

print("README.md créé avec succès")
```
